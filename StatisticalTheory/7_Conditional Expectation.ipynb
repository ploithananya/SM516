## ค่าคาดหมายแบบมีเงื่อนไข(ConditionalExpectation)


การคาดการณ์(prediction)หรือการประมาณค่า(estimation)ที่ได้รับความสนใจมักจะ อ่ในรูปของค่าคาดหมายแบบมีเงื่อนไข(conditional expectation)

- การคาดการณ์(prediction)หรือการประมาณค่า(estimation)มักจะต้องเริ่มจากการที่“รู้ อะไรบางอย่าง”แล้วจึงพยายามที่จะบอกว่าจะเกิดอะไรขึ้นกับตัวแปรหรือสิ่งที่สนใจ
-  ยกตัวอย่างเช่น
    - รายได้ของครัวเรือนที่หัวหน้าครัวเรือนเรียนจบปริญญาตรีแตกต่างจากครัวเรือนที่หัวหน้าครัว เรือนจบอาชีวศึกษานั้นมีค่าเท่าใด
    -  อัตราผลตอบแทนของกองทุนจะเป็นเท่าใดหากอัตราผลตอบแทนของตลาดเท่ากับ10 เปอร์เซ็นต์ต่อปี

### นิยาม: ค่าคาดหมายแบบมีเงื่อนไข(ConditionalExpectation) 

กำหนดให้ XและY เป็นตัวแปรสุ่มโดยที่ค่าคาดหมายของYมีค่าจำกัดนั่นคือ $\mu Y < \infty $
สำหรับกรณีของตัวแปรสุ่มต่อเนื่องค่าคาดหมายแบบมีเงื่อนไข(conditionalexpectation)ของYเมื่อทราบค่าของX = x เท่ากับ
<br><br><center>$E[Y|X = x]$ = $\int_{-\infty }^{\infty } yf2 (y|x)dy \;\;(1)$
<br><br><div align="left">และค่าคาดหมายแบบมีเงื่อนไข(conditionalexpectation)ของXเมื่อทราบค่าของY = yเท่ากับ 
<br><br><center>$E[X|Y = y]$= $\int_{-\infty }^{\infty } xf1 (x|y)dx \;\;(2)$
<br><br><div align="left">สำหรับกรณีของตัวแปรสุ่มไม่ต่อเนื่องค่าคาดหมายแบบมีเงื่อนไข(conditionalexpectation)ของYเมื่อทราบค่าของX = x เท่ากับ 
<br><br><center>$ E[Y|X = x] =∑_y yf2 (y|x)dy \;\;(3)$ 
<br><br><div align="left">และค่าคาดหมายแบบมีเงื่อนไข(conditionalexpectation)ของXเมื่อทราบค่าของY = yเท่ากับ
<br><br><center>$ E[X|Y = x] =∑_x xf1 (y|x)dy \;\;(4)$ 

### สรุป: ค่าคาดหมายแบบมีเงื่อนไข(ConditionalExpectation)
- ค่าคาดหมายแบบมีเงื่อนไข(conditionalexpectation)ของYเมื่อทราบค่าของ $\;\;$ X = x  $\;\;$ ซึ่งแทนด้วย $E[Y|X = x]$ นั่นจะถูกพิจารณาเป็น**"ค่าคงที่ค่าหนึ่ง"**
- แต่หากนำเอาค่าคาดหมายแบบมีเงื่อนไขสำหรับค่าต่างๆของxมารวบรวมเข้าใน“รูปแบบ ของการแจกแจง(distribution)”ซึ่งเขียนในรูป $E[Y|X]$ จะได้ผลลัพธ์ออกมาเป็นตัวแปร สุ่มแบบหนึ่ง 
- โดยสรุป
    <br><br><center> $E[Y|X = x]$ เป็นค่าคงที่
    <br><br><center> $E[Y|X]$ เป็นตัวแปรสุ่ม


### นิยาม: ค่าคาดหมายแบบมีเงื่อนไข(ConditionalExpectation)ของY เมื่อทราบX
- กำหนดให้ $h(x) = E[Y|X = x]$ เป็นฟังก์ชันของ x นิยามให้ $E[Y|X] ≡h(X)$ เป็นค่าคาด หมายแบบมีเงื่อนไข(conditional expectation)ของ$\;$Y$\;$เมื่อทราบ$\;$X
    - $E[Y|X]$ เป็นฟังก์ชันของตัวแปรสุ่ม $\,$ X $\;$ ดังนั้นจึงเป็นตัวแปรสุ่มเช่นกัน

## กฏการหาค่าคาดหมายซ้ำ(Law of Iterative Expectation)
- ทฤษฎีบทต่อไปนี้แสดงวิธีการคำนวณหา”ค่าคาดหมายของตัวแปรสุ่มจากค่าคาดหมาย แบบมีเงื่อนไข”**โดยอาศัยหลักการที่ว่าค่าคาดหมายแบบมีเงื่อนไข $E[Y|X]$ เป็นตัวแปรสุ่มที่ ขึ้นอยู่กับตัวแปรสุ่มที่เป็นเงื่อนไขซึ่งในที่นี้หมายถึงตัวแปรสุ่มX**
- หากเราดำเนินการหาค่าคาดหมายโดยใช้การแจกแจงของตัวแปรสุ่ม $\;\; $X$\;\;$ อีกครั้งหนึ่งก็จะได้ค่าคาดหมายของX(ไม่ใช่ตัวแปรสุ่มอีกต่อไป) 
- หลักการนี้มีประโยชน์อย่างมากในการวิเคราะห์เชิงเศรษฐศาสตร์และการเงินซึ่งมักเรียก คุณสมบัตินี้ว่า**“กฏการหาค่าคาดหมายซ้ำ(Law of Iterative Expectation)”**
- ในขณะที่นักสถิติเรียกทฤษฎีบทนี้ว่ากฏของความน่าจะเป็นรวมสำหรับค่าคาดหมาย(Law of Total Probability for Expectations)

### ทฤษฎี: กฏการหาค่าคาดหมายซ้ำ(Law of Iterative Expectation)
กำหนดให้ X และ Y เป็นตัวแปรสุ่มโดยที่ค่าคาดหมายของ Y มีค่าจำกัดนั่นคือ $\mu Y < \infty $ ดังนั้น 
$E[E[Y|X]] = E[Y]$
เพื่อประหยัดพื้นที่ขอนำเสนอเฉพาะการพิสูจน์สำหรับตัวแปรสุ่มต่อเนื่องดังต่อไปนี้
<br><br><center>$E[E[Y|X]] =\int_{-\infty }^{\infty }E[Y|x]f1 (x)dx = \int_{-\infty }^{\infty }\int_{-\infty }^{\infty } yg2 (y|x)f1 (x)dx$ 
<br><br><div align="left">เนื่องจาก $g2 (y|x)f1 (x) = f(x,y)$ ทำให้สรุปได้ว่า 
<br><br><center>$E[E[Y|X]] =\int_{-\infty }^{\infty }\int_{-\infty }^{\infty } yf(x,y)dx = E[Y]$

### ตัวอย่าง: การหาค่าคาดหมายโดยประยุกต์ใช้กฏการหาค่าคาดหมายซ้ำ

สมมุติว่าXและYเป็นตัวแปรสุ่มที่มีการแจกแจงเอกรูป (uniform distribution) พิจารณาการสุ่มที่เริ่มจากการสุ่มเลือกค่า x จาก การแจกแจงเอกรูปที่มีค่าอยู่ในช่วง [0,1] หลังจากนั้นจึงสุ่มเลือกค่าyจากการแจกแจงเอกรูปที่มีค่าอยู่ในช่วง [x,1] นั่นคือการ แจกแจงแบบมีเงื่อนไขของYเมื่อทราบค่า X = x เป็นการแจกแจงเอกรูปที่มีค่าอยู่ในช่วง [x,1] คำถามก็คือค่าคาดหมาย (expectation) ของYมีค่าเท่าใด?
- ค่าคาดหมายแบบมีเงื่อนไขของYเมื่อทราบค่าX = xมีค่าเท่ากับ 
<br><br><center>$E[Y|X = x] = \frac{1+x}{2} \Rightarrow E[Y|X] = \frac{1+X}{2}$
- <div align="left">ดังนั้นเมื่อประยุกต์ใช้กฏการหาค่าคาดหมายซ้ำ(Law of Iterative Expectation)เพื่อหาค่าคาดหมาย(expectation)ของ Y จะได้ว่า 
<br><br><center>$ E[Y] = E[E[Y|X]] = E[\frac{1+X}{2} ]=\frac{1+\frac{1}{2}}{2}= \frac{3}{4}$
<br><br><div align="left">ทั้งนี้สมการที่สามประยุกต์ใช้ผลลัพธ์ที่ว่าค่าคาดหมายของตัวแปรสุ่มที่มีการแจกแจงเอกรูปในช่วง [0,1] เท่ากับ $E[X] = \frac{1}{2}$

### ตัวอย่าง: การหาค่าคาดหมายโดยประยุกต์ใช้กฏการหาค่าคาดหมายซ้ำ เมื่อทราบว่า $E[Y|X] = aX+b$

<br><br><center>$E[XY] E[XY] = E[E[XY|X]] = E[aX^2 +bX]= aE[X^2]+bE[X]$

### ตัวอย่าง: การหาค่าคาดหมายโดยใช้ค่าคาดหมายแบบมีเงื่อนไข

สมมุติให้การทดลองแต่ละครั้งเป็นอิสระต่อกันและมีความน่าจะเป็นที่การทดลองนี้จะสำเร็จใน แต่ละครั้งมีค่าเท่ากับ p ถ้าอยากจะทดลองให้สำเร็จติดต่อกัน k ครั้งคำถามคือเราน่าจะทดลองกี่ครั้ง กำหนดให้ $N_k$ คือจำนวนการทดลองที่จะพบความสำเร็จติดต่อกัน $k$ ครั้ง กำหนดให้ $M_k$ คือ ค่าคาดหวังของจำนวนการทดลองที่มีความสำเร็จติดต่อกัน $k$ ครั้ง 
<br><br><center>$M_k = E[N_k] = E[E[N_k|N_{k−1}]] \;\;(6) $
<br><br><div align="left">เมื่อกำหนดให้ค่าคาดหวังของจำนวนการทดลองที่จะพบความสำเร็จติดต่อกัน $k$ ครั้ง,$(N_k)$ เมื่อการทดลองก่อนหน้าสำเร็จคือ <br><br><center>$E[N_k|N_{k−1}] = p(N_k−1 + 1) + (1−p)(N_{k−1} + 1 +M_k)$ 
<br><br><center>$E[N_k|N_{k−1}] = N_k−1 + 1 + (1−p)M_k\;\;(7) $ 

### ตัวอย่าง: การหาค่าคาดหมายโดยใช้ค่าคาดหมายแบบมีเงื่อนไข(ต่อเนื่อง)
ประยุกต์ใช้กฏการหาค่าคาดหมายซ้ำ(Law of Iterative Expectation) เพื่อคำนวณหาค่า จาก(7)

<br><br><center>$M_k = M_{k−1} + 1 + (1−p)M_k$
<br><br><center>$M_k = \frac{1}{p} + \frac{M_{k−1}}{p} \;\; (8)$   
<br><br><div align="left">เราทราบด้วยว่าค่าคาดหวังของจำนวนการทดลองที่จะพบความสำเร็จ 1 ครั้งคือ
<br><br><center>$E[N_1] = M_1 = \frac{1}{p} \;\; (9)$

### ข้อสังเกตค่าคาดหมายแบบมีเงื่อนไข (Conditional Expectation) 
- ค่าคาดหมายแบบมีเงื่อนไขมีความสัมพันธ์แบบเชิงเส้น (linear relationship) และทราบค่า คาดหมาย(mean) $ \mu X$,$\mu Y$ ค่าความแปรปรวน(variance) $\sigma^2_X,\sigma^2_Y$ และค่าสหสัมพันธ์ (correlation) $/rho$ ของตัวแปรสุ่มทั้งสองตัว X และ Y ก็จะสามารถระบุค่าคงที่ a และ b ใน รูปค่าสถิติดังกล่าวและเขียนค่าคาดหมายแบบมีเงื่อนไขได้เป็น
<br><br><center> $E[Y|X] = \mu Y + \rho \sigma Y \sigma X (X−\mu X) \;\;(11)$
<br><br><div align="left">
- รูปแบบความสัมพันธ์นี้ที่ได้มีความลักษณะคล้ายคลึงกับความสัมพันธ์ที่ได้จากการประมาณ ค่าแบบกําลังสองน้อยที่สุด(Ordinary Least Square หรือที่เรียกสั้นๆว่าOLS)ถัดๆไป $\mu $โดยจะเป็นได้่าค่าสัมประสิทธิ์ของ X นั้นขึ้นอยู่กับค่าสหสัมพันธ์เป็นหลักโดยเฉพาะอย่าง ยิ่งในการกำหนดว่าค่าจะเป็นบวกหรือลบ
- ส่วนต่อไปนำเสนอความแปรปรวนแบบมีเงื่อนไข(conditional variance)ซึ่งใช้หลักการ เดียวกับ่าคาดหมายแบบมีเงื่อนไขหรือในมุมหนึ่งก็คือค่าคาดหมายแบบมีเงื่อนไขของ ฟังก์ชันของตัวแปรสุ่มนั่นเอง

## นิยามของค่าคาดหมายและค่ามัธยฐานในฐานะตัวทำนาย (Mean and Median as Predictors) 
กำหนดให้ Y เป็นตัวแปรสุ่ม และ d คือตัวทำนาย(predictor)ดังนั้น ค่าคาดหมายของค่าคลาดเคลื่อนกำลังสอง(m.s.e.) ของตัวทำนายdนิยามได้เป็น
<br><center> $E[(Y−d)^2] $
<br><div align="left">ทฤษฎีบทต่อไปนี้แสดงให้เห็นว่าค่าคาดหมาย(mean)คือตัวทำนายที่ทำให้ค่าคาดหมายของค่าคลาดเคลื่อนกำลังสอง(m.s.e.) มีค่าต่ำที่สุด 
กำหนดให้ X เป็นตัวแปรสุ่มที่ ค่าความแปรปรวนมีค่าจำกัดและกำหนดให้ $\mu = E[Y]$ ดังนั้น สำหรับค่าจำนวนจริง d ใดๆ 
<br><center>$E[(Y−d)^2] \geq  E[(Y−\mu)^2]$
<br><div align="left">ซึ่งจะเป็นจริงในรูปสมการก็ต่อเมื่อ d = $\mu$ 

### การพิสูจน์ทฤษฎีบทคือตัวทำนายที่ทำให้ค่าคาดหมายของค่าคลาดเคลื่อน กำลังสอง(m.s.e.) มีค่าต่ำที่สุด 

เราสามารถเขียนค่าคาดหมายของค่าคลาดเคลื่อนกำลังสอง(m.s.e.) ในรูปฟังก์ชันของตัวทำนาย d ได้เป็น

<br><center>$E[(Y−d)^2]= E[Y^2 −2dY+d^2]= E[Y^2]−2d\mu +d^2$
<br><br><div align="left">ซึ่งเป็นฟังก์ชันกำลังสอง(quadraticfunction)
เราสามารถประยุกต์ใช้หลักการแคลคูลัสพื้นฐานในการหาค่าต่ำสุดได้ด้วยการกำหนดให้ตัวทำนายที่ทำให้ได้ค่าฟังก์ชันนี้มีค่าต่ำสุด $d^{*}$ คือค่าที่ทำให้ค่าอนุพันธ์อันดับที่หนึ่งมีค่าเท่ากับศูนย์นั่นคือ
<br><br><center>$\frac{\partial }{\partial d}(E[Y^2]−2d^{∗}\mu + (d^{∗})^2)= 0 ⇒−2\mu + 2d^{∗} = 0 ⇒d^{∗} = \mu$
<br><br><div align="left">เนื่องจากฟังก์ชันจุดประสงค์(objective function) เป็นแบบกำลังสองค่าทำให้ได้ค่าต่ำสุดจึงมีค่าเดียว(unique)ผลที่ตาม มาก็คือสำหรับค่าจำนวนจริง d ใดๆ 
<br><center>$E[(Y−d)^2] \geq  E[(Y−\mu)^2]$

### ค่าคาดหมายและค่ามัธยฐานในฐานะตัวทำนาย(Mean and Median as Predictors)

- ทฤษฎีบทต่อไปนี้แสดงให้เห็นว่าข้อสรุปแบบเดียวกันนี้เป็นจริงสำหรับกรณีที่ใช้ค่าคาดหมายแบบมีเงื่อนไข(conditional mean) แทนค่าคาดหมายธรรมดา 
- สมมุติว่ามีตัวแปรสุ่มอีกตัวหนึ่ง X ซึ่งมีความสัมพันธ์กับตัวแปรสุ่มที่ต้องการทราบหรือ ทำนายค่า Y และที่สำคัญเนื่องจากทราบค่าของXตัวทำนายจึงอยู่ในรูปของฟังก์ชันของX นั่นคือd(X) 
- ตัวทำนายที่ทำให้ค่าคาดหมายของค่าคลาดเคลื่อนกำลังสอง(m.s.e.) มีค่าต่ำที่สุดก็คือ“ ค่าคาดหมายแบบมีเงื่อนไข(conditional mean)ของ Y เมื่อทราบ X”


### ทฤษฎีที่ว่าตัวทำนาย (predictor) คือ ค่าคาดหมายแบบมีเงื่อนไข (conditional mean) 

กำหนดให้ X และ Y เป็นตัวแปรสุ่มที่ค่าความแปรปรวนมีค่าจำกัดตัวทำนาย(predictor)ของตัวแปรสุ่ม Y เมื่อทราบค่า X ที่ทำให้ค่าคาดหมายของค่าคลาดเคลื่อนกำลังสอง(m.s.e.) $E[(Y−d(X))^2]$ มี่าต่ำที่สุดคือ $d^{∗}(X) = E[Y|X]$ 

กำหนดให้ $d(X)$ คือตัวทำนายใดๆซึ่งเป็นฟังก์ชันของ X และ $d^{∗} (X) = E[Y|X]$ สิ่งที่ต้องการพิสูจน์คือ 
<br><br><center>$E[(Y−d^{∗} (X))^2]\leq E[(Y−d(X))^2]$ 
<br><br><div align="left">ซึ่งทำได้ด้วยการประยุกต์ใช้กฏการหาค่าคาดหมายซ้ำ(Law of Iterative Expectation) ซึ่งทำให้สามารถเขียนค่าคาดหมายของค่าคลาดเคลื่อนกำลังสอง(m.s.e.) สำหรับตัวทำนาย $d(X)$ ใดๆในรูปของค่าคาดหมายแบบมีเงื่อนไข(conditional expectation)ได้เป็น  
<br><center>$E[(Y−d(X))^2]= E[E[(Y−d(X))^2|X]]$ 

### ทฤษฎีที่ว่าตัวทำนาย (predictor) คือค่าคาดหมายแบบมีเงื่อนไข (conditional mean)Con’t 

ซึ่งบอกเป็นนัยว่าสิ่งที่ต้องการพิสูจน์ คือ
<br><br><center>$E[E[(Y−d^{∗} (X))^2 |X]]\leq E[E[(Y−d(X))^2 |X]]$
<br><br><div align="left">ความสัมพันธ์นี้จะเป็นจริงถ้า
<br><br><center>$E[(Y−d^{∗} (X))^2 |X = x]\leq E[(Y−d(X))^2 |X = x]$

<br><br><div align="left">สำหรับค่า x ใดๆเมื่อนำเอาอสมการที่ (17) ไปเปรียบเทียบกับอสมการที่ (16) จะเห็นได้ว่าอสมการทั้งสองนั้นสมมูล(equivalent)กันหากพิจารณาว่าตัวทำนายใน $E[(Y−d(X))^2 |X = x]$ นั้นเป็นค่าจำนวนจริงเช่นเดียวกับ d ในอสมการที่ (16) สรุปได้ว่า 
<br><br><center>$E[(Y−E[Y|X])^2 |X = x] \leq E[(Y−d(X))^2 |X = x]$
<br><br><div align="left">สำหรับฟังก์ชัน $d(X)$ และค่าจำนวนจริง x ใดๆดังนั้นจึงสามารถสรุปได้ว่า 
<br><br><center> $E[E[(Y−d^{∗} (X))^2 |X]]\leq E[E[(Y−d(X))^2 |X]]$
<br><br><div align="left">ซึ่งนำไปสู่บทสรุปของทฤษฎีบทที่ต้องการ
