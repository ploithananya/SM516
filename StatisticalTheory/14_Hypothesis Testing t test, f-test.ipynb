# <center>การทดสอบสมมุติฐาน(Hypothesis Testing)

## บทนำ

 - หัวข้อนี้นำเสนอหลักการและวิธีการที่เกี่ยวข้องกับการทดสองสมมุติฐาน โดยให้ความสำคัญกับหลักการต่อไปนี้
      - การทดสอบสมมุติฐานด้วยค่าสถิติที่มีการแจกแจงแบบที (t-test)
      - การทดสอบสมมุติฐานด้วยค่าสถิติที่มีการแจกแจงแบบเอฟ (F-test)

## หลักการสำคัญ: เราทดสอบด้วยการแจกแจงแบบทีเพราะค่าสถิติที่ใช้มีการแจกแจงแบบที

 - ประเด็นสำคัญคือ การแจกแจงของค่าสถิติที่ใช้ในการทดสอบสมมุติฐานขึ้นกับรูปแบบของค่าสถิติ ไม่ใช่การจดจำว่าการทดสอบสมมุติฐานแบบใดใช้การแจกแจงแบบที
 - หากค่าสถิติอยู่ในรูปแบบที่ตัวส่วนมีการแจกแจงแบบปกติมาตรฐานและตัวหารเท่ากับรากที่สองของตัวแปรที่มีการแจกแจงแบบไค(Chi-square Distribution) ค่าสถิตินั้นก็จะมีการแจกแจงแบบที
 - การทดสอบที่ใช้ค่าสถิตินั้นต้องอ้างอิงกับการแจกแจงแบบที

## การทดสอบสมมุติฐานทางเลือกด้านเดียว(One-side Alternative) โดยที่ไม่ทราบค่าความแปรปรวน

 - พิจารณาสมมุติฐานด้านเดียว(One-side Hypothesis) ที่เกี่ยวข้องกับค่าคาดหมาย $\mu$ ของการแจกแจงแบบปกติที่มีพารามิเตอร์ $(\mu,\sigma^2)$ โดยที่ไม่ทราบว่าค่าความแปรปรวนที่แท้จริง $\sigma^2$ มีค่าเท่าใด 
 
 $$H_0:\mu\le\mu_0 \ และ \ H_1:\mu >\mu_0$$<div align ='right'> (1)</div>
 
 - เรามักนิยมใช้ค่าสถิติ 
 
 $$U=\frac{\sqrt{n}(\bar{X}_n-\mu_0)}{\hat{\sigma}}$$<div align ='right'> (2)</div>
 
 
>โดยที่ $\bar{X}_n = \frac{\sum_{i=1}^{n}X_i}{n}$ หมายถึงค่าคาดหมายตัวอย่าง(Sample Mean) และ $\hat{\sigma} = \sqrt{\frac{\sum_{i=1}^{n}(X_i-\bar{X}_n)^2}{n-1}}$ หมายถึงตัวประมาณค่าของส่วนเบี่ยงเบนมาตรฐาน(Standard Deviation Estimator)

 - ประเด็นสำคัญก็คือว่า ถ้า $\mu = \mu_0$ ตัวส่วน $\bar{X}_n\sim (\mu_0,\frac{\sigma^2}{n})$ และ $\frac{\sum_{i=1}^{n}(X_i-\bar{X}_n)^2}{\sigma^2}\sim \chi^2_{n-1}$

## การทดสอบสมมุติฐานทางเลือกด้านเดียว(One-side Alternative) โดยที่ไม่ทราบค่าความแปรปรวน

  - เราสามารถเขียนค่าสถิติ $U$ ในรูปแบบใหม่ได้เป็น
  
  $$U =\frac{\frac{\bar{X}_n-\mu_0}{\frac{\sigma}{\sqrt{n}}}}{\frac{\hat{\sigma}}{\sigma}} = \frac{\frac{\bar{X}_n-\mu_0}{\frac{\sigma}{\sqrt{n}}}}{\sqrt{\frac{\frac{\sum_{i=1}^{n}(X_i-\bar{X}_n)^2}{\sigma^2}}{n-1}}}= \frac{Z}{\sqrt{\frac{Y}{n-1}}}\sim t_{n-1},$$<div align ='right'> (3)</div>
  
 - ประเด็นสำคัญ: $Z$ มีการแจกแจงปกติมาตรฐานและ $Y$ มีการแจกแจงแบบ $\chi^{2}_{n-1}$
 - ดังนั้น $U$ มีการแจกแจงแบบทีที่ระดับความอิสระ(d.o.f) $n-1$
 - กระบวนการทดสอบสมมุติฐาน(1) ที่ใช้ค่าสถิติ $U$ มักจะอยู่ในรูปแบบที่ว่า จะปฏิเสธ $H_0$ ถ้า $U\ge c$ สำหรับค่าวิกฤต $c$ ที่กำหนด คำถามต่อมาก็คือ ฟังก์ชันอำนาจทางสถิติ(Power Function) ของกระบวนการทดสอบนี้มีคุณสมบัติอย่างไรบ้าง?

## ฟังก์ชันอำนาจทางสถิติ(Power Function) สำหรับ $H_0:\mu\le\mu_0$

**Theorem**

>กำหนดให้ $X_1,...,X_n$ เป็นตัวอย่างสุ่มที่สุ่มเลือกมาจากการแจกแจง $N(\mu,\sigma^2)$ และกำหนดให้ $U$ เป็นค่าสถิติที่สอดคล้องกับสมการ(2) และแทนค่าวิกฤตสำหรับการทดสอบที่ระดับนัยสำคัญ $\alpha_0$ ด้วย $c = T^{-1}_{n-1}(1-\alpha_0)$ โดยที่ $T_{n-1}(t)$ คือฟังก์ชันความหนาแน่นสะสมของการแจกแจงแบบทีที่ระดับความอิสระ(d.o.f.) $n-1$ ดังนั้น ฟังก์ชันอำนาจทางสถิติ(Power Function) $\pi(\mu,\sigma^2|\delta)$ สำหรับกระบวนการทดสอบ $\delta$ ที่ทดสอบสมมุติฐาน $H_0:\mu\le \mu_0$ ซึ่งจะปฏิเสธ $H_0$ ถ้า $U\ge c$ มีคุณสมบัติต่อไปนี้

 - $\pi(\mu,\sigma^2|\delta)=\alpha_0$ เมื่อ $\mu = \mu_0$
 - $\pi(\mu,\sigma^2|\delta)<\alpha_0$ เมื่อ $\mu < \mu_0$
 - $\pi(\mu,\sigma^2|\delta)>\alpha_0$ เมื่อ $\mu > \mu_0$
 - $\pi(\mu,\sigma^2|\delta)\rightarrow 0$ เมื่อ $\mu \rightarrow  -\infty$
 - $\pi(\mu,\sigma^2|\delta)\rightarrow 1$ เมื่อ $\mu \rightarrow  \infty$
 
>และกระบวนการทดสอบนี้มีขนาดการทดสอบเท่ากับ $\alpha_0$

## พิสูจน์: การทดสอบสมมุติฐานทางเลือกด้านเดียว(One-side Alternative) โดยที่ไม่ทราบค่าความแปรปรวน

**Proof**

 - ส่วนนี้เป็นการพิสูจน์คุณสมบัติที่ว่า $\mu = \mu_0$ และ $pi(\mu,\sigma^2|\delta)=\alpha_0$ การที่ทราบว่า$\mu = \mu_0$ และทราบว่า $\sigma$ ช่วยให้สามารถสรุปได้ว่า $Z=\frac{\bar{X}_n-\mu_0}{\frac{\sigma}{\sqrt{n}}}\sim N(0,1)$ และ $Y= \frac{\sum_{i=1}^{n}(X_i-\bar{X}_n)^2}{\sigma^2}\sim \chi^2_n-1$ ดังนั้น ในกรณีนี้ค่าสถิติ $U\sim t_{n-1}$ ดังนั้น ฟังก์ชันอำนาจทางสถิติ(Power Function) 
 
 $$\pi(\mu,\sigma^2|\delta) = Pr(U\ge c|\mu_0,\sigma^2) = 1-T_{n-1}(T^{-1}_{n-1}(1-\alpha_0)) = \alpha_0$$<div align ='right'> (4)</div>
 
 
 - สำหรับการพิสูจน์คุณสมบัติที่ว่า $\pi(\mu,\sigma^2|\delta)<\alpha_0$ เมื่อ $\mu<\mu_0$ เราจะเริ่มจากการแบ่ง่สวน $U=U^{*}-W^{*}$ โดยที่ 
 
 $$U^{*} = \frac{\sqrt{n}(\bar{X}_n-\mu)}{\hat{\sigma}}\sim t_{n-1}$$<div align ='right'> (5)</div>
 
 $$W^{*} = \frac{(\sqrt{n}(\mu_0-\mu)}{\hat{\sigma}}$$<div align ='right'> (6)</div>
 
>ซึ่งจะเห็นได้ว่า เราไม่สามารถรุปได้ว่า $U$ มีการแจกแจงแบบที เพราะว่า $E[\bar{X}_n-\mu_0]\neq0$ ภายใต้เงื่อนไขที่ $\mu<\mu_0$ นอกจากนี้ เงื่อนไขดังกล่าวยังทำให้สรุปได้ว่า $W^{*}>0$ เราสามารถคำนวณฟังก์ชันอำนาจทางสถิติ(Power Function) ได้เป็น 

$$\pi(\mu,\sigma^2|\delta) = Pr(U\ge c|\mu_0,\sigma^2) = Pr(U^{*}-W^{*}\ge c|\mu_0,\sigma^2)$$
$$ =Pr(U^{*}\ge c+W^{*}|\mu_0,\sigma^2) < Pr(U^{*}\ge c|\mu_0,\sigma^2)=\alpha_0$$<div align ='right'> (7)</div>

## พิสูจน์: การทดสอบสมมุติฐานทางเลือกด้านเดียว(One-side Alternative) โดยที่ไม่ทราบค่าความแปรปรวน

**Proof**

 - การพิสูจน์คุณสมบัติที่ว่า $\pi(\mu,\sigma^2|\delta)>\alpha_0$ เมื่อ $\mu>\mu_0$ คล้ายคลึงกับการพิสูจน์ในข้อที่ 2 เพียงแต่ค่า $W^{*}<0$ ดังนั้นข้อสรุปที่ได้จึงกลับด้าน
 - การพิสูจน์คุณสมบัติที่ว่า $\pi(\mu,\sigma^2|\delta)\rightarrow\alpha_0$ เมื่อ $\mu\rightarrow\ -\infty$ อาศัยการแยกส่วน $U=U^{*}-W^{*}$ โดยใช้ข้อสรุปที่ว่า $W^{*}\rightarrow \infty$ เมื่อ $\mu\rightarrow -\infty$ ดังนั้น 
 
 $$\pi(\mu,\sigma^2|\delta) = Pr(U\ge c|\mu_0,\sigma^2) = Pr(U^{*}\ge c+W^{*}|\mu_0,\sigma^2) \rightarrow 0 $$
 
>เนื่องจากความน่าจะเป้นที่ตัวแปรสุ่มใดๆ จะมีค่ามากกว่าหรือเท่ากับ $\infty$ เท่ากับศูนย์ 

## พิสูจน์: การทดสอบสมมุติฐานทางเลือกด้านเดียว(One-side Alternative) โดยที่ไม่ทราบค่าความแปรปรวน

**Proof**

 - การพิสูจน์คุณสมบัติที่ว่า  $\pi(\mu,\sigma^2|\delta)\rightarrow 1$ เมื่อ $\mu\rightarrow\ \infty$ อาศัยการแยกส่วน $U=U^{*}-W^{*}$ โดยใช้ข้อสรุปที่ว่า $W^{*}\rightarrow -\infty$ เมื่อ $\mu\rightarrow \infty$ ดังนั้น 
 
 $$\pi(\mu,\sigma^2|\delta) = Pr(U\ge c|\mu_0,\sigma^2) = Pr(U^{*}\ge c+W^{*}|\mu_0,\sigma^2) \rightarrow 1 $$

>เนื่องจากความน่าจะเป้นที่ตัวแปรสุ่มใดๆ จะมีค่ามากกว่าหรือเท่ากับ $-\infty$ เท่ากับหนึ่ง
>นอกจากนี้ คุณสมบัติที่ $1$ และ $2$ รวมกันหมายความว่า สำหรับ $\mu \in \Omega_0$ ใดๆ 

$$\pi(\mu,\sigma^2|\delta) \le \alpha_0 \Rightarrow\alpha_0 = sub_{\mu \in \Omega_0}\pi(\mu,\sigma^2|\delta)$$<div align ='right'> (8)</div>

>ซึ่งก็คือขนาดการทดลองนั่นเอง

## ฟังก์ชันอำนาจทางสถิติ(Power Function) สำหรับ $H_0:\mu\ge\mu_0$

**Theorem**

>กำหนดให้ $X_1,...,X_n$ เป็นตัวอย่างสุ่มที่สุ่มเลือกมาจากการแจกแจง $N(\mu,\sigma^2)$ และกำหนดให้ $U$ เป็นค่าสถิติที่สอดคล้องกับสมการ(2) และแทนค่าวิกฤตสำหรับการทดสอบที่ระดับนัยสำคัญ $\alpha_0$ ด้วย $c = T^{-1}_{n-1}(\alpha_0)$ ดังนั้น ฟังก์ชันอำนาจทางสถิติ(Power Function) $\pi(\mu,\sigma^2|\delta)$ สำหรับกระบวนการทดสอบ $\delta$ ที่ทดสอบสมมุติฐาน $H_0:\mu\ge \mu_0$ ซึ่งจะปฏิเสธ $H_0$ ถ้า $U\le c$ มีคุณสมบัติต่อไปนี้

 - $\pi(\mu,\sigma^2|\delta)=\alpha_0$ เมื่อ $\mu = \mu_0$
 - $\pi(\mu,\sigma^2|\delta)<\alpha_0$ เมื่อ $\mu < \mu_0$
 - $\pi(\mu,\sigma^2|\delta)>\alpha_0$ เมื่อ $\mu > \mu_0$
 - $\pi(\mu,\sigma^2|\delta)\rightarrow 1$ เมื่อ $\mu \rightarrow  -\infty$
 - $\pi(\mu,\sigma^2|\delta)\rightarrow 0$ เมื่อ $\mu \rightarrow  \infty$
 
>และกระบวนการทดสอบนี้มีขนาดการทดสอบเท่ากับ $\alpha_0$

## ทฤษฎี: p-value ของฟังก์ชันความหนาแน่นสะสมของการแจกแจงแบบทีที่ระดับความอิสระ(d.o.f.) 
## n-1

>ทฤษฎีบทต่อไปนี้นำเสนอสูตรสำหรับการคำนวณค่าความน่าจะเป็นที่จะปฎิเสธสมมุติฐานจริง(p-value) สำหรับการทดสอบสมมุติฐานด้วยค่าสถิติที่มีการแจกแจงแบบที

**Theorem**
>กำหนดให้ $u$ คือค่าสถิติ $U$ ที่คำนวณได้จากข้อมุลที่มีอยู่ตามสมการที่ (2) และ $T_{n-1}(t)$ คือ ฟังก์ชันความหนาแน่นสะสมของการแจกแจงแบบทีที่ระดับความอิสระ(d.o.f.) $n-1$ ดังนั้น
 - p=value สำหรับสมมุติฐาน $H_0:\mu\le\mu_0$ เท่ากับ $p = 1-T_{n-1}(u)$
 - p=value สำหรับสมมุติฐาน $H_0:\mu\ge\mu_0$ เท่ากับ $p = T_{n-1}(u)$

## p-value ของฟังก์ชันความหนาแน่นสะสมของการแจกแจงแบบทีที่ระดับความอิสระ(d.o.f.) n-1

**Proof**
>พิจารณาสำหรับสมมุติฐาน $H_0:\mu\ge\mu_0$ เราจะปฏิเสธ $H_0$ ที่ระดับนัยสำคัญ $\alpha_0$ ก็ต่อเมื่อค่าสถิติที่คำนวณได้จากข้อมูลที่มีอยู่ 

$$u\ge T^{-1}_{n-1}(1-\alpha_0) \Rightarrow T_{n-1}(u)\ge 1-\alpha_0 \Rightarrow \alpha_0\ge 1-T_{n-1}(u)$$

>ซึ่งอาศัยคุณสมบัติการเพิ่มขึ้นของฟังก์ชัน $T_{n-1}(u)$ ดังนั้น ค่าระดับนัยสำคัญที่ต่ำที่สุดที่ยังสามารถปฏิเสธ $H_0$ จึงมีค่าเท่ากับ $1-T_{n-1}(u)$ เราสามารถประยุกต์ใช้เหตุผลในทำนองเดียวกันเพื่อพิสูจน์ว่า p-value สำหรับสมมุติฐาน $H_0:\mu\ge\mu_0$ เท่ากับ $p = T_{n-1}(u)$

## ตัวอย่าง: การทดสอบสมมุติฐานทางเลือกด้านเดียว(One-side Alternative) โดยที่ไม่ทราบค่าความแปรปรวน

 - พิจารณาผลตอบแทนสุทธิรายเดือนของกองทุนรวม K-USXNDQ, R, ซึ่งมีการแจกแจงแบบปกติที่มีค่าคาดหมาย $\mu$ และค่าความแปรปรวน $\sigma^2$
 - สมมุติฐานที่ต้องการทดสองคือ $H_0:\mu\ge 0.01$ และ $H_1:\mu < 0.01$
 - การทดสอบในครั้งนี้จะใช้การทดสอบสมมุติฐานโดยใช้ค่าสถิติ
 
 $$U=\frac{\sqrt{n}(\bar{R}_n-0.01)}{\hat{\sigma}},$$
 
 - จากข้อมูลจำนวน $n=24$ เดือน ระหว่างมค. 2561 ถึงเดือนธค. 2562 ค่าเฉลี่ยผลตอบแทนสุทิรายเดือน $\bar{r}_n = 0.01093$ และค่าประมาณของส่วนเบี่ยงเบนมาตรฐาน $\hat{\sigma} = 0.00246$ ดังนั้น ค่าสถิติจากข้อมูลชุดนี้มีค่าเท่ากับ $U=1.847$
 
 - ดังนั้น เราไม่สามารถปฏิเสธสมมุติฐาน $H_0$ ที่ระดับนัยสำคัญ $\alpha_0 = 0.05$ เพราะ 
 
 $$u > T^{-1}_23(0.05) = 1.714$$
 
 - นอกจากนี้ เราสามารถคำนวณหาค่า p-value สำหรับกรณีนี้ได้จาก 
 
 $$p=T_{n-1}(u) = T_{23}(1.847) = 0.9612$$
 

## ตัวอย่าง: การทดสอบสมมุติฐานทางเลือกด้านเดี่ยว (one-sided alternative) โดยที่ไม่ทราบค่าความแปรปรวน

* หากเราสนใจทดสอบสมมุติฐาน $H_0 : \mu \leq 0.01 และ H_1 : \mu < 0.01 แทน$
* ซึ่งใช้ค่าสถิติอันเดียวกันคือ $$U = \frac{\sqrt[]{n}(\bar{R_n}-0.01)}{\hat{\sigma}},$$
* จากข้อมูลจำนวน $n = 24$ เดือน ระหว่างมค. 2561 ถึงเดือนธค. 2562 ค่าเฉลี่ยผลตอบแทนสุทธิรายเดือน $\bar{r_n}=0.01093$ และค่าประมาณของส่วนเบี่ยงเบนมาตรฐาน $\hat{\sigma} = 0.00246 ดังนั้น ค่าสถิติจากข้อมูลชุดนี้มีค่าเท่ากับ U  = 1.847$
* ดังนั้น เราปฏิเสธสมมติฐาน $H_0$ ที่ระดับนัยสำคัญ $\alpha_0 = 0.05 เพราะ u > T_{23}^{-1}  (0.95) = 1.714$
* นอกจากนี้ เราสามารถคำนวณหาค่า p-value สำหรับกรณีนี้ได้จาก $p = 1-T_{n-1}(u) = 1-T_{23}(1.847) = 0.0388$

## การแจกแจงแบบทีที่ตรงกลางไม่เท่ากับศูนย์ (noncentral t distribution

* การคำนวณหาค่าฟังก์ชนอำนาจทางสถิติ (power function) สำหรับสมมุติฐานด้านเดี่ยว(one-sided hypothesis) นั้นทำได้ค่อนข้างลำบาก (ยกเว้นในกรณีของ์   $\mu  = \mu_0)$เนื่องจาก ค่าคาดหมายของ $\bar{X_n} − \mu_0$ ไม่ได้เท่ากับศูนย์สำหรับทุกค่าของ์ $\mu \in \Omega_0$  (ยกเว้นในกรณีของ $\mu    = \mu_0)$
* ทำให้ต้องใช้การแจกแจงแบบทีที่ตรงกลางไม่เท่ากับศูนย์ (noncentral t distribution)
* กำหนดให้ $W และ Y$ เป็นตัวแปรสุ่มที่เป็นอิสระต่อกัน โดยที่ $W ∼ N(\psi,1)และ Y∼X^2_m $ดังนั้น ตัวแปรสุ่ม $$X=\frac{W}{\sqrt[]{\frac{Y}{m}}}$$ <div align ='right'> (9)</div><br>มีการแจกแจงแบบทีที่ตรงกลางไม่เท่ากับศูนย์ (noncentral t distribution) ด้วยระดับความีอิสระ (d.o.f) m และค่าตรงกลางต่างจากศูนย์ (noncentral parameter) $\psi$
* กำหนดให้ $T_m(t|\psi) = Pr(X\leq t)$ แทนฟังก์ชันความหนาแน่นสะสมของการแจกแจงนี้

## ฟังก์ชนอำนาจทางสถิติ (power function) และการแจกแจงแบบทีที่ตรงกลางไม่เท่ากับศูนย์ (noncentral t distribution)

**Theoram**
กำหนดให้ $X_1,...,X_n$ เป็นตัวอย่างสั่มที่มาจาก $N(\mu,\sigma^2)$ดังนั้นการแจกแจงของ $$U = \frac{\sqrt[]{n}(\bar{X_n}-\mu_0)}{\hat\sigma}$$,<div align ='right'> (10)</div>
เป็นแบบทีที่ตรงกลางไม่เท่ากับศูนย์ (noncentral t distribution) ด้วยระดับความีอิสระ (d.o.f) n − 1และค่าตรงกลางต่างจากศูนย์ (noncentral parameter) $\psi = \frac{\sqrt[]{n}(\mu-\mu_0)}{\sigma}$ และ<br> 
* ฟังก์ชนอำนาจทางสถิติสำหรับสมมุติฐาน $H_0 : \mu \leq \mu_0 ที่ปฏิเสธ H_0 เมื่อ U \geq c เท่ากับ$ $$\pi (\mu,\sigma^2 |\delta) = 1-T_{n-1}(c|\psi)$$<div align ='right'> (11)</div>
* ฟังก์ชนอำนาจทางสถิติสำหรับสมมุติฐาน $H_0 : \mu \geq \mu_0 ที่ปฏิเสธ H_0 เมื่อ U \leq c เท่ากับ$ $$\pi (\mu,\sigma^2 |\delta) = T_{n-1}(c|\psi)$$<div align ='right'> (12)</div>

## ตัวอย่าง: การเลือกขนาดของกลุ่มตัวอย่างโดยใช้ฟังก์ชนอำนาจทางสถิติ (power function)

* พิจารณาสมมุติฐาน $H_0 : \mu \leq \mu_0 ที่ปฏิเสธ H_0 เมื่อ U \geq c$
* กำหนดให้ระดับนัยสำคัญ $\alpha_0 = 0.05 และตัวอย่างจำนวน n = 15 ดังนั้น c=T_{14}^{-1}(0.95) = 1.761$
* ค่าฟังก์ชนอำนาจทางสถิติในกรณีนี้ได้จาก $1-T_{n-1}(1.761|\psi)$
* ค่าอำนาจทางสถิติสำหรับค่า $\mu = \mu_0 + \frac{\sigma}{2}$ (สนใจค่าที่หากจากเป้าหมาย 0.5 เท่าของค่าเบี่ยงเบนมาตรฐาน) ได้โดยเริ่มจากการคำนวณหาค่า<br>
$\psi = \frac{\sqrt[]{n}(\mu-\mu_0)}{\sigma} = \frac{\sqrt[]{15}(\mu_0 + \frac{\sigma}{2}-\mu_0)}{\sigma} = \frac{\sqrt[]{15}}{2} = 1.936$
* ดังนั้น ค่าอำนาจทางสถิติในกรณีนี้เท่ากับ<br>
$\pi(\mu = \mu_0 + \frac{\sigma}{2},\sigma^2|\delta) = 1-T_{14}(1.761|1.936) = 1-0.422 = 0.578$

## ตัวอย่าง: การเลือกขนาดของกลุ่มตัวอย่างโดยใช้ฟังก์ชนอำนาจทางสถิติ (power function)

* เราสามารถหาขนาดของตัวอย่างn ที่ทำให้ได้ค่าอำนาจทางสถิติ (power) ณ ค่าพารามิเตอร์  $\mu = \mu_0 + \frac{\sigma}{2} เท่ากับ \bar{\pi} = 0.8 โดยแก้สมการต่อไปนี้$ $$\bar{\pi} = 1-T_{n-1}(T_{n-1}^{-1}(1-\alpha_0)|\frac{\sqrt[]{n}(\mu-\mu_0)}{\sigma}) = 0.8$$
* การประมาณค่า : เนื่องจาก $\alpha_0 = 0.05$ ดังนั้น $$\bar{\pi} = 1-T_{n-1}(T_{n-1}^{-1}(0.95)|\frac{\sqrt[]{n}}{2}) = 0.8$$
* สมมุติว่า ต้องการค่าอำนาจทางสถิติสำหรับค่า เท่ากับ $\bar{\pi}= 0.8 หาค่า n$ ด้วยวิธี trial and error ดังต่อไปนี้
     - ถ้า $n = 26 จะได้ว่า 1-T_{25}(T_{25}^{-1}(0.95)|\frac{\sqrt[]{25}}{2}) = 0.7982$ ซึ่งยังต่ำกว่าเป้าหมาย
     - ทดลองใช้ $n = 27 จะได้ว่า 1-T_{26}(T_{26}^{-1}(0.95)|\frac{\sqrt[]{27}}{2}) = 0.8117$
     - เราควรมีกลุ่มตัวอย่างของเราควรมีขนาดไม่น้อยกว่า $n  = 27$

## การทดสอบสมมุติฐานทางเลือกสองด้าน (two-sided alternative) โดยที่ไม่ทราบค่าความแปรปรวน

* พิจารณาสมุติฐาน $$H_0 : \mu = \mu_0 และ H_1 : \mu \neq \mu_0$$<div align ='right'> (13)</div>
* ค่าสถิติที่นิยมใช้ในการทดสอบ $$U = \frac{\sqrt[]{n}(\bar{X_n}-\mu_0)}{\hat{\sigma}},$$<div align ='right'> (14)</div>
* การทดสอบจะปฏิเสธ $H_0 ถ้า |U| \geq c$
* เนื่องจาภายใต้สมุติฐานว่า $H_0 : \mu = \mu_0 ค่าสถิติ U$ จะมีการแจกแจงแบบทีที่ตรงกลางเท่ากับศูนย์
* ดังนั้น สำหรับการทดสอบที่ต้องการระดับนัยสำคัญ $\alpha_0$ ด้วยกลุ่มตัวอย่าง n ตัวอย่าง ค่าคงที่ $c = T_{n-1}^{-1}(1-\frac{\alpha_0}{2})$

## ตัวอย่าง: การทดสอบสมมุติฐานทางเลือกสองด้าน (two-sided alternative) โดยที่ไม่ทราบค่าความแปรปรวน

* ตัวอย่างนี้ทดสอบสมมุติฐาน $$H_0 : \mu = 0.01 และ H_1 : \mu \neq 0.01$$
โดยใช้ข้อมูลชุดเดี่ยวกันกับตัวอย่างก่อนหน้านี้
* ค่าเฉลี่ยของผลตอบแทนสุทธิรายเดือนของกองทุนรวม K-USXNDQ $\bar{r_n} = 0.010930$ และ ค่าประมาณของส่วนเบี่ยงเบนมาตรฐาน $\hat{\sigma} = 0.00246$
* ภายใต้สมุติฐานว่า $H_0 : \mu = 0.01$ ค่าสถิติ $U = \frac{\sqrt[]{n}(\bar{r_n}-0.01)}{\hat{\sigma}}$ จะมีการแจกแจงแบบทีที่ตรงกลางเท่ากับศูนย์
* ดังนั้น ค่าวิกฤติสำหรับการทดสอบที่ระดับนัยสำคัญ $\alpha_0 = 0.05$ เท่ากับ $c  = T_{23}^{-1}(0.975) = 2.0687$ ส่วนค่าสถิติ $u  = 1.847$ ดังนั้น เราไม่สามารถปฏิเสธ สมมุติฐาน $H_0$ ที่ระดับนัยสำคัญ $\alpha_0 = 0.05$

## ฟังก์ชนอำนาจทางสถิติ (power function) และp-value สำหรับการทดสอบสมมุติฐานทางเลือกสองด้าน (two-sided alternative)

* ฟังก์ชนอำนาจทางสถิติ (power function) สำหรับค่าพารามิเตอร์ที่แท้จริง $\mu_1$ สำหรับการทดสอบสอบสมมุติฐาน $H_0 : \mu   = \mu_0$
$$\pi(\mu,\sigma^2|\delta) = T_{n-1}(-c|\psi)+1-T_{n-1}(c|\psi)$$<div align ='right'> (15)</div>
โดยที่ $c = T_{n-1}^{-1}(1-\frac{\alpha_0}{2}) และ \psi = \frac{\sqrt[]{n}(\mu-\mu_0)}{\sigma}$
* ค่าความน่าจะเป็นที่จะปฏิเสธสมมุติฐานจริง (p-value) สำหรับการทดสอบสมมุติฐาน $H_0 : \mu  = \mu_0$ ที่ระดับนัยสำคัญ $\alpha_0 เท่ากับ p  = 2 [1 − T_{n−1}$ (|u| โดยที่u  คือค่าสถิติ U ที่คำนวณได้จากข้อมูลที่มีอยู่ตามสมการที่(2) และ$T_{n−1} (t)$ คือฟังก์ชันความหนาแน่นสะสมของการแจกแจงแบบทีที่ระดับความีอิสระ (d.o.f)$ n − 1$

## การทดสอบสมมุติฐาน $H_0 : \sigma^2_{1} = \sigma^2_{2}$: F-test

* พิจารณาสมมุติฐาน 
$$H_0 : \sigma^2_{X} = \sigma^2_{Y} และ H_1 : \sigma^2_{X} \neq \sigma^2_{Y}$$
* ค่าสถิติที่นิยมใช้ในการทดสอบสสมมุติฐานแบบนี้คือ
$$U = \frac{\frac{S_x^2}{m-1}}{\frac{S_Y^2}{n-1}}$$
โดยที่ $S_X^2 = \sqrt[]{\frac{\sum_{i=1}^n (X_i-\bar{X_n})^2}{n-1}} และ S_Y^2 = \sqrt[]{\frac{\sum_{i=1}^n (Y_i-\bar{Y_n})^2}{n-1}}$ และ m และ n คือขนาดตัวอย่างของ X และY ตามลำดับ
* ประเด็นสำคัญก็คือถ้า X และY มีการแจกแจงแบบปกติ
$$\frac{S_X^2}{\sigma_X^2}∼X_{m-1}^2 และ\frac{S_Y^2}{\sigma_Y^2}∼X_{n-1}^2 \Rightarrow \frac{\frac{S_x^2}{\sigma_X^2(m-1)}}{\frac{S_Y^2}{\sigma_Y^2(n-1)}}∼ F_{m-1,n-1} $$

## การทดสอบสมมุติฐาน $H_0 : \sigma^2_{1} = \sigma^2_{2}$: F-test

* ภายใต้สมมุติฐาน $H_0 : \sigma^2_{1} = \sigma^2_{2}$ เราจึ่งสามารถสรุปได้ว่า
$$\frac{\frac{S_x^2}{(m-1)}}{\frac{S_Y^2}{(n-1)}}∼ F_{m-1,n-1}$$
* ประเด็นสำคัญก็คือเราจะใช้การแจกแจงแบบใดในการทดสอบสมมุติฐานอันใด ก็ขึ้นอยู่กับว่า เราใช้ค่าสถิติแบบใดในการทดสอบสมมุติฐานนั้น
