### <span style=color:purple>รูปตัวอย่างของการแจกแจงปกติ (Normal Distribution)</span>

### <span style=color:purple>การแจกแจงปกติมาตรฐาน (Standard Normal Distribution Function)</span>

* การแจกแจงปกติมาตรฐาน (Standard Normal Distribution Function)
    * การแจกแจงปกติที่มีความพิเศษคือ **“การแจกแจงปกติมาตรฐาน (standard normal distribution)”** ซึ่งมีฟังก์ชันความหนาแน่นของความน่าจะเป็น (p.d.f.) เท่ากับ

$$
\Phi(z) = \frac{1}{\sqrt2\Pi}{e^{-z^{\frac{2}{2}}}}
$$

* **“ฟังก์ชันก่อกำเนิดโมเมนต์(m.g.f.) ของตัวแปรสุ่ม Z”** ที่มีการแจกแจงปกติมาตรฐาน (standard normal distribution) เท่ากับ

$$
\begin{align} 
{\varphi_{z}(t)\&=E[{e^{tz}}]} &= \int_{-\infty }^{\infty}e^{tz}\frac{1}{\sqrt2\Pi}{e^{-z^{\frac{2}{2}}}}dz
\\
&= \int_{-\infty }^{\infty}e^{{\frac{t^{2}}{2}}}\frac{1}{\sqrt2\Pi}e^{-(z^{2}-2tz+t^{2})}dz
\\
&= \int_{-\infty }^{\infty}e^{{\frac{t^{2}}{2}}}\frac{1}{\sqrt2\Pi}e^{-\frac{(z-2)^{2}}{2}}dz
\\
&= e^{t^{\frac{2}{2}}}\int_{-\infty }^{\infty}\frac{1}{\sqrt2\Pi}e^\frac{\tilde{z}^{2}}{2}d\tilde{z}
\\
&=e^{{\frac{t^{2}}{2}}}
\end{align}
$$
โดยที่ในขั้นตอนก่อนสุดท้ายใช้เทคนิคการเปลี่ยนตัวแปร $\tilde{z}$ = z - t

### <span style=color:purple> ค่าความคาดหมาย (mean) และค่าความแปรปรวน (variance) ของการแจกแจงปกติมาตรฐาน (standard normal distribution)</span>

* ค่าความคาดหมาย (mean) และค่าความแปรปรวน (variance) หาได้จาก ฟังก์ชันก่อกำเนิดโมเมนต์(m.g.f.) จากทฤษฎีบทที่ก่อนหน้า

$$
\begin{align} 
E[z] &=  \frac{d}{dt}\psi _{z}(t)\mid_{t=0}
\\
&=  \frac{d}{dt}e^{{\frac{t^{2}}{2}}}\mid_{t=0}
\\
&= te^{{\frac{t^{2}}{2}}}\mid_{t=0}
\\
&= 0
\end{align}
$$

$$
\begin{align} 
Var[Z] &= E[Z^{2}] 
\\
&= \frac{ d^{2}}{dt^{2}}e^{\frac{t^{2}}{2}}\mid_{t=0}  
\\
&= [e^{\frac{t^{2}}{2}}+t^{2}e^{\frac{t^{2}}{2}}]\mid_{t=0} 
\\
&= 1
\end{align}
$$

* **การแจกแจงปกติมาตรฐาน (standard normal distribution)** คือ การแจกแจงแบบปกติที่มี**“ค่าความคาดหมายเท่ากับศูนย์และค่าความแปรปรวนเท่ากับหนึ่ง”** ซึ่งมักเขียนแทนการแจกแจงแบบนี้ด้วย N (0,1)

### <span style=color:purple>ฟังก์ชันเชิงเส้นของการแจกแจงปกติ(Normal Distribution Function)</span>

* เราสามารถสร้างตัวแปรสุ่ม X ที่มีการการแจกแจงปกติ(normal distribution) ที่มีค่าความคาดหมาย (mean) เท่ากับ $\mu $ และค่าความแปรปรวน (variance) เท่ากับ $\sigma ^{2}$ ได้จากตัวแปรสุ่ม Z โดยใช้ความสัมพันธ์เชิงเส้นต่อไปนี้

$$
X = \sigma Z + \mu 
$$

* ดังนั้น ฟังก์ชันก่อกำเนิดโมเมนต์(m.g.f.) ของตัวแปรสุ่ม X เท่ากับ

$$
\begin{align} 
\psi _{x}(t) &= e^{\mu t}e^\frac{{(\sigma t)^{2}}}{2}
\\
&= e^{\mu t+\frac{1}{2}\sigma^{2} t^{2}} 
\end{align}
$$

* ค่าความคาดหมาย (mean) และค่าความแปรปรวน (variance) คือ

$$
\begin{align} 
E[X] &= \frac{\mathrm{d} }{\mathrm{d} t}\psi _{x}(t)\mid_{t=0} 
\\
&= (\mu +t\sigma^{2} )e^{\mu t+\frac{1}{2}\sigma^{2}t^{2} }\mid_{t=0}
\\
&= \mu
\end{align}
$$

$$
\begin{align} 
Var[X] &= E[X^{2}] - \mu ^{2}
\\
&= \frac{d^{2}}{dt^{2}}e^{\mu t+\frac{1}{2}\sigma^{2}t ^{2}}\mid_{t=0}-\mu^{2} 
\\
&= [\sigma^{2}e ^{\mu t+\frac{1}{2}\sigma^{2}t ^{2}}+(\mu +t\sigma^{2} )^{2}e ^{\mu t+\frac{1}{2}\sigma^{2}t ^{2}}]\mid_{t=0}
\\
&= \sigma ^{2}+\mu ^{2}-\mu ^{2}
\\
&= \sigma ^{2}
\end{align}
$$

* ตัวแปรสุ่ม X = $\sigma$Z + $\mu$ มีการการแจกแจงปกติ(normal distribution) ที่ค่าความคาดหมาย (mean) เท่ากับ $\mu$ และค่าความแปรปรวน (variance) เท่ากับ $\sigma^2$ ซึ่งมีฟังก์ชันความหนาแน่นของความน่าจะเป็น (p.d.f.) เท่ากับ

$$
f(x) = \frac{1}{\sqrt{2\pi \sigma^{2} }}e^{\frac{-(x-\mu )^{2}}{2\sigma^{2} }}
$$  

*  และสามารถเขียนแทนการแจกแจงแบบนี้ด้วย N($\mu$ ,$\sigma^{2}$ )

* หากทราบว่าตัวแปรสุ่ม X มีการการแจกแจงปกติ(normal distribution) ที่ค่าความคาดหมาย (mean) เท่ากับ $\mu$ และค่าความแปรปรวน (variance) เท่ากับ $\sigma ^{2}$ $\neq$ 0 แล้วก็จะทราบทันทีว่าตัวแปรสุ่ม Z = $\frac{X-\mu }{\sigma }$ จะมีการแจกแจงปกติมาตรฐาน (standard normal distribution)

* การแปลงค่ารูปแบบนี้คือการปรับคะแนนมาตรฐาน (standardized score) ซึ่งนิยมใช้ในการปรับค่าผลลัพธ์หรือผลการทดสอบให้อยู่ในรูปของตัวแปรที่มีการแจกแจงปกติมาตรฐาน

### <span style=color:purple>ทฤษฎี: ความสมมาตรของการแจกแจงปกติ</span>

### <span style=color:purple>Theorem (ความสมมาตรของการแจกแจงปกติ)</span>

สำหรับค่าจำนวนจริง z ใดๆ

$$
\Phi (-z) = 1 - \Phi(z)
$$

และสำหรับ 0 < p < 1

$$
\Phi^{-1}(p) = -\Phi^{-1}(1-p)
$$

### <span style=color:purple> การพิสูจน์: ความสมมาตรของการแจกแจงปกติ </span>

<span style=color:purple>Proof</span>

* เริ่มจากความสัมพันธ์ที่ว่า

$$
Pr(Z\leq -z) + Pr(Z> -z) = 1
$$

* เนื่องจากการแจกแจงปกติเป็นการแจกแจงที่สมมาตร (symmetric) ทำให้  $Pr(Z>-z) = Pr(Z\leq z)$ ดังนั้น

$$
Pr(Z\leq -z) + Pr(Z\leq z) = 1 \Rightarrow = \Phi(-z) = 1-\Phi(z)
$$

* สมการที่8 พิสูจน์ได้ด้วยการแทนค่า $z = \Phi^{-1}(p)$ และส่วนกลับของมัน $\Phi(z)=p$ ลงในด้านซ้ายและด้านขวาของสมการ ตามลำดับ

$$
\Phi(-\Phi^{-1}(p)) = 1-p \Rightarrow -\Phi(p)=\Phi^{-1}(1-p)
$$

* สอดคล้องกับสมการ $\Rightarrow$ [$\Phi^{-1}(p) = -\Phi^{-1}(1-p)$]

### <span style=color:purple> ทฤษฎี: ความสัมพันธ์ระหว่างการแจกแจงปกติและการแจกแจงปกติมาตรฐาน</span>

<span style=color:purple>Theorem</span>

กำหนดให้X เป็นตัวแปรสุ่มที่มีแจกแจงปกติ(normal distribution) ด้วยค่าคาดหมาย (mean) $\mu$ และค่าความแปรปรวน (variance) $\sigma^2$ กำหนดให้ F แทนฟังก์ชันความน่าจะเป็นสะสม (C.D.F.) ของ X ดังนั้น ตัวแปรสุ่ม $z=\frac{x-\mu }{\sigma }$ มีการแจกแจงปกติมาตรฐาน (standard normal distribution) โดยที่สำหรับค่าจำนวนจริง x ใดๆ

$$
F(x)=\Phi(\frac{x-\mu }{\sigma })
$$

และสำหรับค่าจำนวนจริง 0 < p < 1 ใดๆ

$$
F^{-1}(p)=\mu +\sigma \Phi^{-1}(p)
$$

### <span style=color:blue> ตัวอย่าง: การหาความน่าจะเป็นของการแจกแจงแบบปกติ</span>

#### <span style=color:blue>Example</span>

สมมุติให้ตัวแปรสุ่ม X มีการแจกแจงแบบปกติด้วยด้วยค่าคาดหมาย (mean) $\mu=10$ และค่าความแปรปรวน (variance) $\sigma^2=5$ คำถามก็คือ ความน่าจะเป็นที่X จะมีค่าอยู่ในช่วง 5 < X < 15 มีค่าเท่าใด?

เริ่มจากการแปลงตัวแปรสุ่มให้เป็นแบบปกติมาตรฐานด้วยฟังก์ชัน $z=\frac{x-\mu }{\sigma }$

$$
Pr (5 < X < 15) = Pr(\frac{5-10}{5}) < (\frac{x-10}{5}) < (\frac{15-10}{5}) = Pr(-1 < Z< 1)
$$

ซึ่งสามารถเขียนในรูปของ $\Phi$ ได้เป็น

$$
\begin{align} 
Pr (-1 < Z < 1) &= Pr(\left | Z \right |< 1)
\\
&= \Phi(1)-\Phi(-1)
\approx 0.6826
\end{align}
$$

### <span style=color:blue>ทฤษฎี: ฟังก์ชันเชิงเส้น (linear) ของตัวแปรสุ่มปกติจะยังมีการแจกแจงปกติ</span>

#### <span style=color:blue> Theorem (ฟังก์ชันเชิงเส้น (linear) ของตัวแปรสุ่มปกติจะยังมีการแจกแจงปกติ) </span>

ถ้า X เป็นตัวแปรสุ่มที่มีการแจกแจงปกติ(normally distributed) ด้วยค่าคาดหมาย (mean) $\mu$ และค่าความแปรปรวน (variance) $\sigma^2$ และ Y = aX + b สำหรับค่าคงที่a และ b โดยที่ $a \neq 0$ แล้ว Y จะมีการแจกแจงปกติ(normally distributed) ด้วยค่าคาดหมาย (mean) $a\mu +b$ และค่าความแปรปรวน (variance) $a^{2}\sigma ^{2}$

<span style=color:blue>Proof</span>

ฟังก์ชันก่อกำเนิดโมเมนต์(m.g.f.) ของตัวแปรสุ่ม X เท่ากับ

$$
\psi_x(t)=e ^{\mu t+\frac{1}{2}\sigma^{2}t ^{2}}
$$

ดังนั้น ฟังก์ชันก่อกำเนิดโมเมนต์(m.g.f.) ของตัวแปรสุ่ม Y = aX + b เท่ากับ

$$
\begin{align}
\psi_y(t)&=e^{bt}\psi_x(at)
\\
&= e^{bt}e^{\mu at+\frac{1}{2}\sigma^2a^2t^2}
\\
&= e^{(a\mu +b)t+\frac{1}{2}(\sigma^2a^2)t^2}
\end{align}
$$

ซึ่งเป็นฟังก์ชันที่ตรงกับรูปแบบของฟังก์ชันก่อกำเนิดโมเมนต์(m.g.f.) ของการแจกแจงปกติที่มีค่าคาดหมาย (mean) a$\mu$ + b และ ค่าความแปรปรวน (variance) $\sigma^2a^2$

### <span style=color:magenta>ทฤษฎี: การแจกแจงปกติ(normally distributed) ของตัวแปรสุ่มที่เป็นอิสระต่อกัน</span>

<span style=color:magenta>Theorem (ตัวแปรสุ่มที่เป็นอิสระต่อกัน) </span>

ถ้าตัวแปรสุ่ม $X_1,... ,X_n$ เป็นอิสระต่อกัน และ Xi มีการแจกแจงปกติ(normally distributed) ด้วยค่าคาดหมาย (mean) $\mu_i$ และค่าความแปรปรวน (variance) $\sigma^2_i$ สำหรับ $i = 1,...,n$ แล้ว $Y = \sum _{i=1}^{n}a_iX_i+b$ มีการแจกแจงปกติ(normally distributed) ด้วยค่าคาดหมาย (mean) $\sum_{i=1}^{n}a_iX_i+b$ และค่าความแปรปรวน (variance)  $\sum _{i=1}^{n}a^2_i\sigma^2_i$

<span style=color:magenta>Proof</span>

ฟังก์ชันก่อกำเนิดโมเมนต์(m.g.f.) ของ Xi เท่ากับ

$$
\psi_i(t)=e ^{\mu t+\frac{1}{2}\sigma^{2}t ^{2}}
$$

เนื่องจากตัวแปรสุ่มเหล่านี้เป็นอิสระต่อกัน ฟังก์ชันก่อกำเนิดโมเมนต์(m.g.f.) ของ Y เท่ากับ

$$
\begin{align}
\psi _{y}(t)\mid &=\prod_{i=1}^{n}e^{bt}\psi _i(a_it)
\\
&= \prod_{i=1}^{n}e^{bt}e^{\mu _ia_it+\frac{1}{2}a^2_i\sigma^2_it^2}
\\
&= e^{(\sum_{i=1}^{n}a_i\mu_i+b )+\frac{1}{2}(\sum_{i=1}^{n}a^2_i\sigma^2_i)t^2}\
\end{align}
$$

ซึ่งบ่งบอกว่า Y มีการแจกแจงปกติ (normally distributed) ด้วยค่าคาดหมาย (mean) $\sum_{i=1}^{n}a_i\mu_i+b$ และค่าความแปรปรวน (variance) $\sum_{i=1}^{n}a^2_i\sigma^2_i$

### <span style=color:green> นิยาม: ค่าเฉลี่ยของตัวอย่างสุ่ม (random sample)</span>

#### <span style=color:green>Definition</span>

กำหนดให้ $X_1,..., X_n$ เป็นตัวแปรสุ่ม ค่าคาดหมายของตัวอย่าง (sample mean) นิยามได้เป็น

$$
{\bar{x}}_n=\frac{\sum_{i=1}^{n}X_i}{n}
$$

* หากพิจารณาจากข้อมูล (observed data) n ตัวอย่าง ค่าเฉลี่ย (average) นั้นหมายถึง ${\bar{x}}_n=\frac{\sum_{i=1}^{n}X_i}{n}$ โดยที่ $X_i$ คือค่าที่เกิดขั้นจริง (realized value) ของตัวแปรสุ่ม $X_i$ หรือค่าตัวแปรของตัวอย่าง 1,...,n

* ค่าเฉลี่ย (average) ${\bar{x}}_n$ คือสำเนาจากตัวอย่าง (sample counterpart) ของค่าคาดหมายของตัวอย่าง (sample mean) ${\bar{x}}_n$ สิ่งที่แตกต่างกันก็คือ ค่าเฉลี่ย (average) ${\bar{x}}_n$ เป็นจำนวนจริงค่าหนึ่ง **ส่วนค่าคาดหมายของตัวอย่าง (sample mean) ${\bar{x}}_n$ เป็นตัวแปรสุ่ม** 

### <span style=color:purple>ทฤษฎี: ค่าเฉลี่ยของตัวอย่างสุ่ม (random sample) ที่สุ่มมาจากการแจกแจงปกติ(normal distribution) จะมีการแจกแจงปกติ</span>

#### <span style=color:purple>Theorem </span>

สมมุติให้ X_1,...,X_n เป็นตัวแปรสุ่มที่นำไปสู่ตัวอย่างสุ่ม (random sample) ขนาด n และแต่ละตัวเป็นตัวแปรสุ่มที่มีการแจกแจงปกติ(normally distributed) ด้วยค่าคาดหมาย (mean) $\mu$ และค่าความแปรปรวน (variance) $\sigma^2$ ดังนั้น ค่าคาดหมายของตัวอย่าง (sample mean) ${\bar{x}}_n$ เป็นตัวแปรสุ่มที่มีการแจกแจงปกติ(normally distributed) ด้วยค่าคาดหมาย (mean) $\mu$ และค่าความแปรปรวน (variance) $\frac{\sigma ^2}{n}$  นั่นคือ

$$
\bar{x}\sim N(\mu,\frac{\sigma^2}{n})
$$

#### <span style=color:purple>Proof </span>

ทฤษฎีบทนี้เป็นผลมาจากการประยุกต์ใช้ทฤษฎีบทที่แล้วโดยกำหนดให้ $a_i = \frac{1}{n}$ และ b = 0 ซึ่งช่วยให้สรุปได้ว่า ${\bar{x}}_n=\frac{\sum_{i=1}^{n}X_i}{n}$ เป็นตัวแปรสุ่มที่มีการแจกแจงปกติ(normally distributed) ด้วย ค่าคาดหมาย (mean) $\sum_{i=1}^{n}\frac{1}{n}\mu=\mu$ และ ค่าความแปรปรวน (variance) $\sum_{i=1}^{n}(\frac{1}{n})^2\sigma^2=\frac{\sigma^2}{n}$

### <span style=color:purple>ความน่าจะเป็นและ $\sigma$</span>

* การประยุกต์ใช้การแจกแจงปกติ(และการแจกแจงที) ในการทดสอบสมมุติฐานทำได้โดยง่าย  
$\triangleright $ การใช้งานในลักษณะของเกณฑ์มาตรฐาน (benchmark) คือ ที่สองเท่าของค่าเบี่ยงเบนมาตรฐาน 2$\sigma$ จะหมายถึงประมาณร้อยละ 95    
$\triangleright $ สามเท่าของค่าเบี่ยงเบนมาตรฐาน $3\sigma$ จะหมายถึงมากกว่าร้อยละ 99  
$\triangleright $ เกณฑ์มาตรฐานแบบนี้ช่วยให้นักวิเคราะห์สามารถอ่านผลการประมาณค่าได้อย่างรวดเร็วโดยไม่ต้องพึ่งตารางสถิติ  

* การแจกแจงอันหนึ่งที่เชื่อมโยงกับการแจกแจงปกติคือ การแจกแจงปกติด้วยล็อก (lognormal distribution)ซึ่งหมายถึงการแจกแจงของตัวแปรสุ่มที่ล็อกการิธึมของตัวแปรนั้นมีการแจกแจงปกติ(normal distribution)

### <span style=color:purple>นิยาม: การแจกแจงปกติด้วยล็อก (lognormal distribution)</span>

#### <span style=color:purple> Definition</span>

ถ้า ln X มีการแจกแจงปกติ(normal distribution) ด้วยค่าคาดหมาย (mean) $\mu$ และค่าความแปรปรวน (variance) $\sigma^2$ แล้ว ตัวแปรสุ่ม X จะมีการแจกแจงปกติด้วยล็อก (lognormal distribution) ที่มีพารามิเตอร์ $\mu$ และ $\sigma^2$

* ฟังก์ชันความหนาแน่นของความน่าจะเป็น (p.d.f.) ของตัวแปรสุ่ม X ที่มีการแจกแจงปกติด้วยล็อก (lognormal distribution) สำหรับพารามิเตอร์ $\mu$ และ $\sigma^2 $เท่ากับ

$$
\begin{align}
f(x)&=\frac{1}{x\sqrt{2\pi \sigma ^2}}e^{-\frac{(lnx-\mu)^2}{2\sigma^2}}  สำหรับ X > 0   
\\
f(x) &= 0, สำหรับกรณีอื่น
\end{align}
$$

* ฟังก์ชันความน่าจะเป็นสะสม (C.D.F.) เท่ากับ

$$
\begin{align}
F(x)&=Pr(X\leq x)
\\
&= Pr(\frac{lnX-\mu}{\sigma}\leq \frac{lnx-\mu}{\sigma})
\\
&= \Phi(\frac{lnx-\mu}{\sigma})
\end{align}
$$

### <span style=color:purple>ค่าคาดหมาย (mean) และค่าความแปรปรวน (variance)ของการแจกแจงปกติด้วยล็อก (lognormal distribution)</span>

* การหาค่าคาดหมาย (mean) และค่าความแปรปรวน (variance) นั้นมีเทคนิคหนึ่งที่ช่วยให้คำนวณหาค่าได้ง่ายคือ การใช้ฟังก์ชันก่อกำเนิดโมเมนต์(m.g.f.) ของตัวแปรที่มีการแจกแจงปกติซึ่งในที่นี้หมายถึง Y = ln X

$$
\psi _{y}(t)=E[e^{ty}]=e^{\mu t+\frac{1}{2}\sigma^2t^2}
$$

* สมการสุดท้ายเป็นผลมาจากการที่Y = ln X มีการแจกแจงปกติ(normal distribution) ด้วยค่าคาดหมาย (mean) $\mu$ และค่าความแปรปรวน (variance) $\sigma^2$ในขณะเดียวกัน เราสามารถเขียนฟังก์ชันก่อกำเนิดโมเมนต์(m.g.f.) ในรูปของ X ได้เป็น

$$
\psi _{y}(t)=E[e^{ty}]=E[e^{tlnX}]=E[X^t]
$$

* ค่าคาดหมายของ X เท่ากับ

$$
E[X]=\psi_{y}(1)=e ^{\mu +\frac{1}{2}\sigma^2 }
$$

* ค่าความแปรปรวนของ X เท่ากับ

$$
\begin{align}
Var[X]&=E[X^2]-E[X]^{2}
\\
&=\psi_y(2)-\psi_y(1)^{2}
\\
&=e^{2\mu +2\sigma^{2}}-(e^{\mu +\frac{1}{2}\sigma ^{2}})^{2}
\\
&=e^{2\mu +\sigma^{2} }[e^{\sigma^2}-1]
\\
\end{align}
$$

### <span style=color:purple>ตัวอย่าง: การคำนวณราคาอนุพันธ์เพื่อซื้อ (call options)</span>

#### <span style=color:purple> Example</span>

สมมุติให้ราคาหลักทรัพย์ณ เวลา $t,S_t$ มีการแจกแจงปกติด้วยล็อก (lognormal distribution) นั่นคือ $S_t=S_0e^{Z_t}$ โดยที่ $Z_t$ มีการแจกแจงปกติ(normal distribution) ที่มีค่าคาดหมาย (mean) $\mu t$ และค่าความแปรปรวน (variance) $\sigma^2t$

* ความสะดวกในการคำนวณ สามารถเขียน $Z_t$ ในรูปของตัวแปรสุ่มที่มีการแจกแจงปกติมาตรฐาน (standard normal distribution) Z ได้เป็น

$$
Z_t = \mu t+\sigma \sqrt{t}Z
$$

* สามารถเขียนราคาหลักทรัพย์ในรูปการแจกแจงปกติมาตรฐาน (standard normal distribution) Z ได้เป็น

$$
S_t=S_0e^{\mu t+\sigma \sqrt{t}Z}
$$

* พิจารณาอนุพันธ์เพื่อซื้อ (call options) ซึ่งหมายถึงอนุพันธ์ที่ให้สิทธิแก่ผู้ซื้อในการซื้อหลักทรัพย์ที่ราคาอ้างอิง (strike price) K ณ เวลาที่กำหนด T แต่ไม่จำเป็นต้องใช้สิทธิ์นั้นถ้าไม่ต้องการ ดังนั้นมูลค่า (value) ของอนุพันธ์เพื่อซื้อ (call option) ณ เวลาที่กำหนด T มีค่าเท่ากับ

$$
\begin{align}
V(S)&=max\left \{ S-K,0\left.  \right \} \right.
\\
V(S)&=S-K,ถ้า S>K
\\
V(S)&=0,ถ้าเป็นอย่างอื่น
\\
\end{align}
$$

### <span style=color:purple> วิธีทำ: การคำนวณราคาอนุพันธ์เพื่อซื้อ (call options) </span>

#### <span style=color:purple> Example </span>

* การกำหนดราคาอนุพันธ์(option pricing) ในที่นี้จะประยุกต์ใช้หลักการเป็นกลางต่อความเสี่ยง (risk neutrality)

* กำหนดว่า ค่าคาดหมายของมูลค่าของหลักทรัพย์ควรจะมีค่าเท่ากับผลลัพธ์ที่ได้จากการฝากเงินแบบไม่มีความเสี่ยงที่อัตราดอกเบี้ย r นั่นคือ $e^{rt}S_0=E[S_t]$ ซึ่งมีผลทำให้สามารถสรุปได้ว่า

$$
e^{rt}S_0=E[S_0e^{\mu t+\sigma \sqrt{t}Z}]\Rightarrow e^{rt}=e^{\mu t+\frac{\sigma^{2} }{2}t}\Rightarrow \mu =r-\frac{\sigma^2}{2}
$$

* หลักการเป็นกลางต่อความเสี่ยง (risk neutrality) ยังระบุว่าราคาอนุพันธ์เพื่อซื้อ, C (K; T), ควรจะมีค่าเท่ากับค่าคาดหมายของมูลค่าของอนุพันธ์ที่ปรับส่วนลด (discount) ด้วยอัตราดอกเบี้ย r นั่นคือ

$$
\begin{align}
C(K,t)&=e^{-rt}E[V(S_T)]
\\
&=e^{-rt}E[max\left \{ S_t-K,0 \right \}]
\end{align}
$$

* เพื่อจะคำนวณค่าคาดหมายนี้จำเป็นต้องแยกออกเป็นสองกรณีคือ
    * กรณีที่ $S_T$ > K ซึ่งจะเป็นจริงก็ต่อเมื่อ
        $$S_0e^{\mu T+\sigma \sqrt{T}Z}> K\Rightarrow Z> \frac{ln\tfrac{K}{S_0}-\mu T}{\sigma \sqrt{T}}\equiv \bar{z}_{T}$$
    * กรณีที่ $S_T$ > K ซึ่งจะเป็นจริงก็ต่อเมื่อ
        $$ Z< \frac{ln\tfrac{K}{S_0}-\mu T}{\sigma \sqrt{T}}\equiv \bar{z}_{T} $$
        

* ราคาอนุพันธ์เพื่อซื้อ (call options) ที่สามารถใช้สิทธิ์ได้ณ เวลา T ที่ราคาอ้างอิง (strike price) K มีค่าเท่ากับ

$$
\begin{align}
C(K,T)&=e^{-rt}E[max\left \{ S_t-K,0 \right \}]
\\
&=e^{-rt}\int_{\bar{z}_{T} }^{\infty }[S_0e^{\mu t+\sigma \sqrt{T}z}-K]\phi (z)dz
\\
&=e^{-rt+\mu t}S_0\int_{\bar{z}_{T}} ^{\infty }e^{\sigma\sqrt{t}z}\phi(z)dz-Ke^{-rt}\int_{\bar{z}_{T} }^{\infty }\phi (z)dz
\end{align}
$$

### <span style=color:purple>วิธีทำ: การคำนวณราคาอนุพันธ์เพื่อซื้อ (call options), Black-Scholes Model</span>

#### <span style=color:purple> Example </span>

* พิจารณาการอินทิเกรตในพจน์แรกด้านขวา ดังนี้

$$
\begin{align}
\int_{\bar{z}_T}^{\infty }e^{\sigma\sqrt{T}z}\phi(z)dz&=\int_{\bar{z}_T}^{\infty}e^{\sigma\sqrt{T}z}\frac{1}{\sqrt{2\pi }}e^{-\frac{z^2}{2}}dz
\\
&=e^{\frac{\sigma^2}{2}T}\int_{\bar{z}_T}^{\infty }\frac{1}{\sqrt{2\pi }}e^{-\frac{(z-\sigma\sqrt{t})^{2}}{2}}dz
\\
&=e^{\frac{\sigma^2}{2}T}\int_{\bar{z}_T-\sigma\sqrt{T}}^{\infty }\frac{1}{\sqrt{2\pi }}e^{-\frac{Z^2}{2}}dz
\\
&=e^{\frac{\sigma^2}{2}T}[1-\Phi(\bar{z}_T-\sigma\sqrt{T})]
\\
&=e^{\frac{\sigma^2}{2}T}\Phi(\sigma\sqrt{T}-\bar{z}_T)
\end{align}
$$

* เป็นผลมาจากการเปลี่ยนรูปให้เป็นกำลังสองสัมบูรณ์ส่วนสมการที่สามเป็นผลมาจากการเปลี่ยนตัวแปรอินทิเกรตจาก z เป็น $z-\sigma\sqrt{T}$ และ สมการสุดท้ายเป็นผล
มาจากคุณสมบัติความสมมาตรของการแจกแจงปกติมาตรฐาน เมื่อแทนค่ากลับเข้าไปในสมการราคาอนุพันธ์เพื่อซื้อ (call options) จะได้ว่า

$$
\begin{align}
C(K,T)&=e^{-rT+\mu T}S_0e^{\frac{\sigma^2}{2}T}\Phi(\sigma\sqrt{T}-\bar{z}_T)-Ke^{-rt}\Phi(-\bar{z}_T)
\\
&=e^{-rT+\mu T+\frac{\sigma^2}{2}T}\Phi(\sigma\sqrt{T}-\bar{z}_T)-Ke^{-rt}\Phi(-\bar{z}_T)
\\
&=S_0\Phi(\sigma\sqrt{T}-\bar{z}_T)-Ke^{-rt}\Phi(-\bar{z}_T)
\end{align}
$$

เมื่อแทนค่า $\mu=r-\frac{\sigma^2}{2}$ และ $z=\frac{ln\frac{K}{S_0}-(r-\frac{\sigma^2}{2})T}{\sigma\sqrt{T}}$

* สมการนี้คือสมการที่มีชื่อเสียงของ **B|lack-Scholes สำหรับการกำหนดราคาอนุพันธ์ (option pricing)**

### <span style=color:purple>ฟังก์ชันความหนาแน่นของความน่าจะเป็น (p.d.f.) ของการแจกแจงปกติมาตรฐานหลายตัวแปร (Multivariate Normal Distribution)</span>

* สิ่งที่ขาดหายไปจากฟังก์ชันความหนาแน่นความน่าจะเป็น (p.d.f.) ของตัวแปรสุ่มตัวเดียว คือ **ความแปรปรวน (covariance) หรือสหสัมพันธ์(correlation) ซึ่งจะปรากฏในกรณีที่มีตัวแปรหลายตัว**

* เริ่มจากตัวแปรสุ่มปกติมาตรฐานที่เป็นอิสระต่อกัน ซึ่งเขียนแทนด้วยเวคเตอร์สุ่ม (random vector) $Z={[Z_1,...,Z_n]}'$ โดยที่ ${X}'$ หมายถึง เมทริกซ์สลับเปลี่ยน (transpose matrix) ของ X และ $Z_i\sim N(0,1)$ มีการแจกแจงปกติมาตรฐาน (standard normal) และเป็นอิสระต่อกัน(independent)

* นั่นคือตัวแปรสุ่มเหล่านี้มีการแจกแจงเหมือนกันและเป็นอิสระต่อกัน (identically and independently distributed หรือ i.i.d.) ซึ่งในกรณีที่การแจกแจงเป็นแบบปกติมาตรฐาน สามารถเขียนแทนด้วย $Z_i\sim N(0,1_n)$

* ฟังก์ชันความหนาแน่นของความน่าจะเป็น (p.d.f.) ของ Z คือ 

$$
\begin{align}
f(z)&=\prod_{i=1}^{n}\phi(Z_i)
\\
&=\prod_{i=1}^{n}(\frac{1}{2\pi })^{\frac{1}{2}}exp\left \{ -\frac{1}{2}z^2_i \right \}
\\
&=(\frac{1}{2\pi })^{\frac{n}{2}}exp\left \{ -\frac{1}{2} \Sigma_{i=1}^{n}z^2_i\right \}
\\
&=\frac{1}{(2\pi )^{\frac{n}{2}}}exp\left \{ -\frac{1}{2}{z}'z \right \}
\end{align}
$$

### <span style=color:purple>ฟังก์ชันก่อกำเนิดโมเมนต์(m.g.f.) ของการแจกแจงปกติมาตรฐานหลายตัวแปร (Multivariate Normal Distribution)</span>

* สมการข้างต้นเป็นผลมาจากการที่ตัวแปรทั้งหมดมีการแจกแจงปกติมาตรฐานเหมือนกันและเป็นอิสระต่อกัน (i.i.d.)

* ชัดเจนว่า ค่าคาดหมาย (mean) และเมทริกซ์ของค่าความแปรปรวนร่วม (covariance matrix) ของ Z มีค่าเท่ากับ 0 และ $I_n$ ตามลำดับ

* ส่วนฟังก์ชันก่อกำเนิดโมเมนต์(m.g.f.) ของ Z สามารถเขียนได้เป็น

$$
\begin{align}
\psi _z(t)&=E[exp\left \{ {Z}'t \right \}]
\\
&=E\left [ exp\left \{ \Sigma_{i=1}^{n}Z_it_i \right \} \right ]
\\
&=E\left [ \prod_{i=1}^{n}exp\left \{ Z_it_i \right \} \right ]
\\
&=\prod_{i=1}^{n}E[exp\left \{ Z_it_i \right \}]
\\
&=\prod_{i=1}^{n}\psi _z(t_i)
\\
&=\prod_{i=1}^{n}e^{\frac{1}{2}t^2_i}
\\
&=exp\left \{ \frac{1}{2}\Sigma_{i=1}^{n}t^2_i \right \}
\\
&=exp\left \{ \frac{1}{2}{t}' t\right \}
\end{align}
$$

* เราสามารถสร้างเวคเตอร์ของตัวแปรสุ่มปกติ(normal random vector) X ที่มีค่าคาดหมาย (mean) เท่ากับ $\mu$ และเมทริกซ์ของค่าความแปรปรวนร่วม (covariance matrix) เท่ากับ $\Sigma$ 

### <span style=color:purple>การสร้างเวคเตอร์ของตัวแปรสุ่มปกติ(normal random vector) ของการแจกแจงปกติหลายตัวแปร (Multivariate Normal Distribution)</span>

* ในการสร้างเวคเตอร์ของตัวแปรสุ่มปกติ(normal random vector) จะคล้ายกับตัวแปรสุ่มตัวเดียวที่มีการแจกแจงปกติแบบมาตรฐาน แต่มี **ความยุ่งยากทางเทคนิค** ในการนิยามค่าความเบี่ยงเบนในรูปของเมทริกซ์ซึ่งต้องใช้เทคนิคที่เรียกว่า **การแยกส่วนสเปกตรัล (spectral decomposition)**

* เริ่มจากคุณสมบัติที่สำคัญอันหนึ่งของเมทริกซ์ของค่าความแปรปรวนร่วม (variance-covariance matrix) $\Sigma$ คือ มันเป็น **เมทริกซ์จัตุรัสและสมมาตร (symmetric square matrix)** และ **คุณสมบัติเป็นบวกเกือบแน่นอน (positive semidefinite)** ซึ่งทำให้สามารถใช้หลักการทางพีชคณิตเชิงเส้น (linear algebra) แยกส่วน $\Sigma$ ออกได้เป็น

$$
\Sigma ={\Gamma}' \Lambda \Gamma
$$

โดยที่ $\Lambda$ เป็นเมทริกซ์แนวทแยง (diagonal matrix)

$$
\Lambda =\begin{bmatrix}
\lambda_1  &  0  &  ...  &  0  &  0  & \\ 
 0  &  \lambda_2   &  ...  &  0  &  0 \\ 
 \vdots   & \vdots   &  \ddots   &  \vdots  &  \vdots \\ 
 0  &  0  &  ...  &  \lambda_n-1  &  0 \\ 
 0  &  0  &  ...  &  0  &  \lambda_n 
\end{bmatrix}
$$

และ $\Gamma$ เป็นเมทริกซ์ตั้งฉาก (orthogonal matrix) นั่นคือ $\Gamma^{-1}={\Gamma}'$

* สรุปได้ว่า ${\Gamma}'\Gamma= I$ โดยทั่วไปเพื่อความสะดวก เรามักจะเรียงลำดับ \lambda_i จากมากไปหาน้อย นั่นคือ $\lambda_1 \geq \lambda _2\geq \cdots \geq \lambda _n\geq 0$ และเรียก $\lambda_i$ แต่ละตัวว่า ค่าลักษณะเฉพาะ (eigenvalue) ส่วนคอลัมน์ที่i แต่ละอัน (สอดคล้องกับ $\lambda_i$) ซึ่งแทนด้วย $\gamma_i$ มักเรียกว่าเวคเตอร์ลักษณะเฉพาะ (eigenvector) ของ $\sum$

* จากการที่เราสามารถเขียนเมทริกซ์ของค่าความแปรปรวนร่วม (variance-covariance matrix) $\sum$ ในรูปของค่าลักษณะเฉพาะ (eigenvalue) และเวคเตอร์ลักษณะเฉพาะ (eigenvector) ได้ดังนี้

$$
\Sigma ={\Gamma}'\Lambda \Gamma =\Sigma_{i=1}^{n}\lambda _i\gamma _i{\gamma }'_i
$$

* ประโยชน์ที่สำคัญของการแยกส่วนนี้ก็คือ การได้มาซึ่งเมทริกซ์แนวทแยง (diagonal matrix) $\Lambda$ ซึ่งค่าลักษณะเฉพาะ (eigenvalue) $\lambda_i$ แต่ละค่ามีค่าไม่ติดลบ (nonnegative) ซึ่งช่วยให้สามารถนิยามรากที่สอง (square root) ของ $\Lambda$ ได้เป็น

$$
\Lambda ^{\frac{1}{2}}=\begin{bmatrix}
 \sqrt{\lambda_1}  &  0  &  0  &  0  &  0 \\ 
 0  &  \sqrt{\lambda_2}  &  0  &  0  &  0 \\ 
 \vdots   &  \vdots   &  \ddots   &  \vdots  &  \vdots \\ 
 0  &  0  &  ...  &  \sqrt{\lambda_{n-1}}  &  0 \\ 
 0  &  0  &  ...  &  0  &  \sqrt{\lambda_{n}} 
\end{bmatrix}
$$

* ซึ่งมีคุณสมบัติคล้ายกับรากกำลังสองของจำนวนจริง นั่นคือ

$$
\Lambda ^{\frac{1}{2}}\Lambda ^{\frac{1}{2}}=\Lambda 
$$

* เราสามารถเขียนสมการการแยกส่วนสเปกตรัล (spectral decomposition) ใหม่ได้เป็น

$$
\begin{align}
\Sigma &={\Gamma }'\Lambda ^{\frac{1}{2}}\Lambda ^{\frac{1}{2}}\Gamma 
\\
&=({\Gamma }'\Lambda^{\frac{1}{2}}\Gamma  )({\Gamma }'\Lambda^{\frac{1}{2}}\Gamma  )
\end{align}
$$

* ใช้คุณสมบัติการตั้งฉาก (orthogonal) ของเมทริกซ์ $\Gamma$ นั่นคือ $\Gamma{\Gamma }'=I$ ผลที่ตามมาจากสมการดังกล่าวคือ คือ ทำให้สามารถ
นิยามรากที่สอง (square root) ของ $\Sigma$ ได้เป็น

$$
\Sigma{}^{\frac{1}{2}}={\Gamma}' \Lambda ^{\frac{1}{2}}\Gamma
$$

* หากเมทริกซ์ของค่าความแปรปรวนร่วม (variance-covariance matrix) $\Sigma$ เป็นบวกแน่นอน (positive definite หรือ p.d.) แล้วค่าลักษณะเฉพาะ (eigenvalue) มีค่าเป็นบวกทุกค่า นั่นคือ $\lambda_i>0$ สำหรับทุกๆ $i=1,...,n$ ซึ่งมีผลทำให้ค่าดีเทอร์มิแนนต์(determinant) ของ $\Lambda ^{\frac{1}{2}}$ มีค่ามากกว่าศูนย์อย่างแน่นอน ทำให้สามารถหาเมทริกซ์ส่วนกลับ (inverse matrix) ของ $\Lambda ^{\frac{1}{2}}$ ได้ดังนั้นจึงสามารถนิยามเมทริกซ์ส่วนกลับ(inverse matrix) ของ $\Sigma{}^{\frac{1}{2}}$ ได้เป็น

$$
\Sigma{}^{-\frac{1}{2}}={\Gamma}' \Lambda ^{-\frac{1}{2}}\Gamma
$$

โดยที่ $\Lambda ^{-\frac{1}{2}}$ แทนเมทริกซ์ส่วนกลับ (inverse matrix) ของ $\Lambda ^{\frac{1}{2}}$

### <span style=color:blue>ฟังก์ชันก่อกำเนิดโมเมนต์(m.g.f.) ของการแจกแจงปกติหลายตัวแปร (Multivariate Normal Distribution) </span>

* สร้างเวคเตอร์ของตัวแปรสุ่มปกติ(normal random vector) $X \sim N\left (\mu ,\Sigma  \right )$  ที่มีค่าคาดหมาย (mean) เท่ากับ $\mu$ และเมทริกซ์ของค่าความแปรปรวนร่วม (covariance matrix) เท่ากับ $\Sigma$ โดยกำหนดให้

$$
X=\Sigma{}^{-\frac{1}{2}}Z+\mu
$$

* ฟังก์ชันก่อกำเนิดโมเมนต์(m.g.f.) ของ X เท่ากับ

$$
\begin{align}
\psi _x(t)&=E[exp\left \{ {X}'t \right \}]
\\
&=E\left [exp\left \{ {(\Sigma{}^{\frac{1}{2}}Z+\mu)}' \right \}  \right ]
\\
&=exp\left \{ {\mu}'t \right \}E\left [ exp\left \{ {Z}'\left (\Sigma{}^{\frac{1}{2}}t   \right )\right \} \right ]
\\
&=exp\left \{ {\mu}'t  \right \}exp\left \{ \frac{1}{2}{(\Sigma {}^\frac{1}{2}t)}'(\Sigma {}^\frac{1}{2}t) \right \}
\\
&=exp\left \{{\mu }'t+\frac{1}{2}{t}'\Sigma t  \right \}
\end{align}
$$

* ประยุกต์ใช้ฟังก์ชันก่อกำเนิดโมเมนต์(m.g.f.) ของ Z ในสมการที่17 โดยใช้  $\Sigma {}^\frac{1}{2}t$ แทน t สังเกตด้วยว่า ฟังก์ชันก่อกำเนิดโมเมนต์(m.g.f.) ในที่นี้สามารถหาค่าได้ตราบเท่าที่เมทริกซ์ของค่าความแปรปรวนร่วม (variance-covariance matrix) $\Sigma$ เป็นบวกเกือบแน่นอน (p.s.d.) ในขณะที่ฟังก์ชันความหนาแน่นของความน่าจะเป็น (p.d.f.) ของ X จะหาค่าได้ก็ต่อเมื่อ $\Sigma$ เป็นบวกแน่นอน (p.d.) ทั้งนี้เพราะการแปลงฟังก์ชันความหนาแน่นของความน่าจะเป็น (p.d.f.) จำเป็นต้องใช้ค่าดีเทอร์มิแนนต์(determinant) ของ $\Sigma {}^{-\frac{1}{2}}$

### <span style=color:blue>ฟังก์ชันความหนาแน่นของความน่าจะเป็น (p.d.f.) ของการแจกแจงปกติหลายตัวแปร (Multivariate Normal Distribution)</span>

หาก $\Sigma$ เป็นบวกแน่นอน (p.d.) แล้ว จะสามารถเขียนสมการใหม่ได้เป็น

$$
Z=\Sigma {}^{-\frac{1}{2}}(X-\mu)
$$

ดังนั้น ฟังก์ชันความหนาแน่นของความน่าจะเป็น (p.d.f.) ของ X คือ

$$
f(x)=\frac{1}{(2\pi )^{\frac{n}{2}}\left | det\Sigma  \right |^{\frac{1}{2}}}exp\left \{ -\frac{1}{2}{(x-\mu )}' \Sigma {^{-1}}(x-\mu)\right \}
$$

### <span style=color:blue>ทฤษฎี: ความสัมพันธ์เชิงเส้นของการแจกแจงปกติหลายตัวแปร (Multivariate Normal Distribution)</span>

#### <span style=color:blue>Theorem</span>

กำหนดให้X เป็นเวคเตอร์ของตัวแปรสุ่ม n ตัว ซึ่งมีการแจกแจงร่วมแบบปกตินั่นคือ $X\sim N(\mu ,\Sigma )$ และ = AX + b โดยที่A เป็นเมทริกซ์ของค่าคงที่ขนาด m x n และ b เป็นเวคเตอร์ของค่าคงที่ m ตัว แล้ว $Y\sim N\left (A\mu +b,A\Sigma {A}'  \right )$

#### <span style=color:blue>Proof</span>

ฟังก์ชันก่อกำเนิดโมเมนต์(m.g.f.) ของ Y เท่ากับ

$$
\begin{align}
\psi_y(t)&=\left [ exp\left \{ {Y}' t\right \} \right ]
\\
&=E\left [ exp\left \{ {(AX+b)}'t \right \} \right ]
\\
&=exp\left \{ {b}'t \right \}E\left [ exp\left \{{ X}'({A}'t) \right \} \right ]
\\
&=exp\left \{ {b}'t \right \}exp\left \{ {\mu}'\left ( {A }'t\right )+\frac{1}{2}{\left ({A }'t \right ) }'\Sigma\left ( {A}'t \right ) \right \}
\\
&=exp\left \{{\left (A\mu +b\right )}'t + \frac{1}{2}{t}' \left ( A\Sigma {A}'\right ) t \right \}
\end{align}
$$

นั่นคือ$ Y\sim N\left (  A\mu +b,A\Sigma{A }'  \right )$

### <span style=color:purple>ประโยชน์ในการแปลงตัวแปรสุ่มในรูปแบบฟังก์ชันเชิงเส้น (linear transformation)</span>

* ทฤษฎีบทด้านบนยังช่วยให้สามารถหาการแจกแจงของตัวแปรสุ่มเพียงบางส่วนในกรณีที่การแจกแจงร่วมเป็นแบบปกติโดยเริ่มจากการที่สามารถแบ่งเวคเตอร์X ออกได้เป็นสองส่วน คือ

$$
X=\begin{bmatrix}
 X_m \\ 
 X_k 
\end{bmatrix}
$$

โดยที่ $X_m$ คือเวคเตอร์ของตัวแปรสุ่มที่สนใจ โดยในที่นี้กำหนดให้มีขนาด m ในขณะที่ตัวแปรสุ่ม $X_k$ มีขนาด k = n - m

* กำหนดให้b = 0 และ

$$
A=\begin{bmatrix}
I_{mm}  &&  &0_{mk}\\ 
 &  & &\\ 
 0_{km}  &&  &0_{kk}
\end{bmatrix}
$$

จะทำให้ฟังก์ชันที่เกิดจากการแปลงค่าโดยใช้A และ b เท่ากับ

$$
Y=\begin{bmatrix}
 I_{mm} & 0_{mk}\\ 
\end{bmatrix}\begin{bmatrix}
 X_m \\ 
 X_k 
\end{bmatrix}=X_m
$$

### <span style=color:purple>สรุป: การแปลงตัวแปรสุ่มในรูปแบบฟังก์ชันเชิงเส้น (lineartransformation) </span>

* เราสามารถแบ่ง ค่าคาดหมาย (mean) และเมทริกซ์ความแปรปรวนร่วม (variance-covariance matrix) ของ X ได้เป็น

$$
\mu =\begin{bmatrix}
\mu_m
\\ 
\mu_k
\end{bmatrix}
$$

และ

$$
\Sigma =\begin{bmatrix}
 \Sigma_{mm}   &\Sigma_{mk}  
\\
\Sigma_{km}  &\Sigma_{kk}
\end{bmatrix}
$$

โดยที่ $\Sigma_{mk}$ คือเมทริกซ์ความแปรปรวนร่วม (variance-covariance matrix) ระหว่าง $X_m$ และ $X_k$

* ค่าคาดหมาย (mean) และเมทริกซ์ความแปรปรวนร่วม (variance-covariance matrix) ของ Y สามารถแบ่งส่วนได้เป็น

$$
\begin{bmatrix}
I_{mm} & 0_{mk}
\end{bmatrix}\begin{bmatrix}
 \Sigma _{mm} & & \Sigma _{mk} \\ 
 \Sigma _{km} & & \Sigma _{kk}
\end{bmatrix}\begin{bmatrix}
I_{mm}
\\ 
0_{mk}
\end{bmatrix}=\Sigma_{mm} 
$$

* สรุปได้ว่า $Y=X_m\sim N(\mu _m,\Sigma _{mm})$ กล่าวคือ การแจกแจงร่วมเป็นแบบปกติแล้วการแจกแจงของตัวแปรสุ่มบางส่วนก็จะมีการแจกแจงปกติเช่นกัน โดยที่ค่าคาดหมาย (mean) และค่าความแปรปรวน (variance) ของชุดตัวแปรบางส่วนสามารถหาได้จากค่าสถิติของตัวแปรสุ่มทั้งหมดได้ไม่ยากนัก

### <span style=color:purple>ตัวอย่าง: การหาฟังก์ชันความหนาแน่นของความน่าจะเป็นร่วม (p.d.f.) ของ $X_1$ และ $X_2$ </span>

#### <span style=color:purple> Example</span>

พิจารณาตัวแปรสุ่มสองตัว $X_1$ และ $X_2$ ที่มีการแจกแจงปกติ(normally distributed) ด้วยค่าคาดหมาย $(\mu_1,\mu_2)$ และเมทริกซ์ความแปรปรวนร่วม (variance-covariance matrix)

$$
\Sigma =\begin{bmatrix}
 \sigma^{2}_1 & & &\sigma_{12} \\ 
 &  &  & \\ 
 \sigma_{12} & & &\sigma^2_2
\end{bmatrix}
$$

จะเห็นได้ว่า $det\Sigma=\sigma^2_1\sigma^2_2-\sigma^2_{12}=\sigma^2_1\sigma^2_2\left ( 1-\rho^2  \right )$ โดยที่ $\rho$ แทนค่าสหสัมพันธ์ระหว่าง $X_1$ และ $X_2$  ดังนั้น ส่วนกลับ (inverse) ของ $\Sigma$ เท่ากับ

$$
\Sigma^{-1} =\frac{1}{\sigma^2_1\sigma^2_2(1-\rho^2)}\begin{bmatrix}
 \sigma^{2}_2 & & &-\sigma_{12} \\ 
 &  &  & \\ 
 -\sigma_{12} & & &\sigma^2_1
\end{bmatrix}
$$

* สามารถคำนวณได้ว่า

$$
\begin{align}
{(x-\mu )}'\Sigma ^{-1}(x-\mu )&=\left [ x_1-\mu _1,x_2-\mu _2 \right ]\frac{1}{\sigma^2_1\sigma^2_2(1-\rho^2)}\begin{bmatrix}
 \sigma^{2}_2 & & &-\sigma_{12} \\ 
 &  &  & \\ 
 -\sigma_{12} & & &\sigma^2_1
\end{bmatrix}\begin{bmatrix}
 &x_1-\mu _1& \\ 
 &  \\ 
 &x_2-\mu_2&
\end{bmatrix}
\\
&=\frac{1}{1-\rho^2}\left [\left (\frac{x_1-\mu_1}{\sigma_1}\right )^2-2\rho\left ( \frac{x_1-\mu_1}{\sigma_1}\right )\left ( \frac{x_2-\mu_2}{\sigma_2}\right)+ \left ( \frac{x_2-\mu_2}{\sigma_2}\right) ^{2}              \right ]
\end{align}
$$

### <span style=color:purple>วิธีทำ: การหาฟังก์ชันความหนาแน่นของความน่าจะเป็นร่วม (p.d.f.) ของ $X_1$ และ $X_2$ </span>

#### <span style=color:purple>Example </span>

* ฟังก์ชันความหนาแน่นของความน่าจะเป็นร่วม (p.d.f.) ของ $X_1$ และ $X_2$ คือ 

$$
f(x_1,x_2)=\frac{1}{2\pi \sigma _1\sigma _2\sqrt{1-\rho ^2}}e^{-\frac{Q}{2}}
$$

โดยที่

$$
Q=\frac{1}{1-\rho^2}\left [\left (\frac{x_1-\mu_1}{\mu_1}\right)^2-2\rho\left ( \frac{x_1-\mu_1}{\sigma_1}\right )\left ( \frac{x_2-\mu_2}{\sigma_2}\right )+ \left ( \frac{x_2-\mu_2}{\sigma_2}\right )^2\right ]
$$

* ถ้า $\rho=0$ แล้ว ฟังก์ชันความหนาแน่นของความน่าจะเป็นร่วม (p.d.f.) ของ $x_1$ และ $x_2$ คือ

$$
\begin{align}
f(x_1,x_2)&=\frac{1}{2\pi \sigma _1\sigma _2\sqrt{1-\rho^2}}exp\left \{-\frac{1}{2}    \left [\left ( \frac{x_1-\mu_1}{\mu_1}\right)^2+\left ( \frac{x_2-\mu_2}{\sigma^2}\right)^2\right]\right \}
\\
&=\left [\frac{1}{\sqrt{2\pi \sigma ^2_1}} e^{-\frac{1}{2}\left (\frac{x_1-\mu_1}{\sigma_1}\right )^2}\right ]\left [\frac{1}{\sqrt{2\pi \sigma ^2_2}} e^{-\frac{1}{2}\left (\frac{x_2-\mu_2}{\sigma_2}\right )^2}\right ]
\\
&=f_1(x_1)f_2(x_2)
\end{align}
$$

* หมายความว่า $x_1$ $x_2$ เป็นอิสระต่อกัน กล่าวคือ ในกรณีของการแจกแจงปกติ(normal distribution) การไม่มีสหสัมพันธ์(uncorrelated) หมายถึงการเป็นอิสระต่อกัน (independent) ด้วย

### <span style=color:green>ทฤษฎี: ความเป็นอิสระต่อกัน (independent) ของตัวแปรสุ่ม</span>

#### <span style=color:green>Theorem</span>

กำหนดให้ $X\sim N(\mu ,\Sigma)$ และสามารถแยกส่วนได้เป็นสองส่วนคือ $X=\left ( X_m,X_k \right )$ ซึ่งสอดคล้องกับสมการด้านบนดังกล่าว ดังนั้น $X_m$ และ $X_k$ เป็นอิสระต่อกัน (independent) ก็ต่อเมื่อ (if and only if) $\Sigma_{mk}=0$

#### <span style=color:green>Proof</span>

พิจารณาฟังก์ชันก่อกำเนิดโมเมนต์(m.g.f.) ของ X ในรูปของ $t_m$ และ $t_k$

$$
\begin{align}
\psi (t_m,t_k)&=exp\left \{{\mu_m}' t_m+{\mu_k}'t_k+\frac{1}{2}\left [{t_m}'\Sigma_{mm}t_m+{t_k}'\Sigma_{mm}t_k+{t_k}'\Sigma_{mk}t_k+{t_k}'\Sigma_{km}t_m\right]\right\}
\\
&=exp\left \{{\mu _m}' t_m+{\mu _k }' t_k+\frac{1}{2}\left [ t{_m}'\Sigma_{mm}t_m+{t_k}'\Sigma_{mm}t_k\right]\right \}exp\left \{\frac{1}{2}\left [{t_m}'\Sigma_{mk}t_k+{t_k}'\Sigma_{km}t_m\right ]\right\} 
\end{align}
$$

$$
\begin{align}
\psi (t_m,t_k)&=exp\left \{{\mu_m }'t_m+\frac{1}{2}\left [{t_m}'\Sigma_{mm}t_m\right ]\right \}exp\left \{{\mu_k}'t_k+\frac{1}{2}\left [{t_k}' \Sigma_{mm}t_k\right ]\right\}
\\
&=exp\left \{{\mu_m}'t_m+{\mu_k}'t_k+\frac{1}{2}\left [{t_m}'\Sigma_{mm}t_m+{t_k}'\Sigma_{mm}t_k\right]\right\}
\end{align}
$$


$$
exp\left \{\frac{1}{2}\left [{t_m }'\Sigma_{mk}t_k+{t_k}'\Sigma_{km}t_m\right]\right\}=1\Rightarrow {t_m}'\Sigma_{mk}t_k+{t_k}'\Sigma_{km}t_m=0
$$

ซึ่งจะเป็นจริงก็ต่อเมื่อ $\Sigma_{mk}={\Sigma_{km}}'=0$

### <span style=color:green>ทฤษฎี: การแจกแจงแบบมีเงื่อนไข (conditional distribution) เป็นการแจกแจงแบบปกติ(normal)</span>

#### <span style=color:green>Theorem</span>

กำหนดให้ $X\sim N(\mu ,\Sigma)$ และสามารถแยกส่วนได้เป็นสองส่วนคือ $X=(X_m,X_K)$ ซึ่งสอดคล้องกับสมการดังกล่าวด้านบน และสมมุติว่า $\Sigma$ เป็นบวกแน่นอน (positive definite) ดังนั้น การแจกแจงแบบมีเงื่อนไข (conditional distribution) ของ $X_m$ เมื่อทราบ $X_k$ เป็นแบบปกติ(normal) นั่นคือ

$$
X_m|X_k\sim N\left (\mu_m+\Sigma_{mk}\Sigma_{kk}^{-1}\left ( X_k-\mu_k \right ),\Sigma_{mm}-\Sigma_{mk}\Sigma_{kk}^{-1} \Sigma_{km}\right)
$$

### <span style=color:green>การพิสูจน์: การแจกแจงแบบมีเงื่อนไข (conditional distribution) เป็นการแจกแจงแบบปกติ(normal)</span>

#### <span style=color:green> Proof </span>

* กำหนดให้ $W=X_m-\Sigma_{mk}\Sigma_{kk}^{-1}X_k$ และ

$$
\begin{bmatrix}
W
\\ 
X_2
\\ 
\end{bmatrix}=\begin{bmatrix}
 &I_m  &  &   &-\Sigma_{mk}\Sigma_{kk} ^{-1}\\ 
 &  &  &  & &\\ 
 &0  &  &  &I_k
\end{bmatrix}\begin{bmatrix}
X_1
\\ 
X_2
\\ 
\end{bmatrix}
$$

* คำนวณหาเมทริกซ์ของค่าความแปรปรวน (variance-covariance matrix) ได้เป็น

$$
\begin{bmatrix}
 &I_m  &  &  &-\Sigma_{mk}\Sigma_{kk}^{-1}&   \\ 
 &  &  &  &   \\ 
 &0  &  &  &I_k&   
\end{bmatrix}\begin{bmatrix}
 &\Sigma_{mm}  &  &  &\Sigma_{mk}&   \\ 
 &  &  &  &   \\ 
 &\Sigma_{km}  &  &  &\Sigma_{kk} &  
\end{bmatrix}\begin{bmatrix}
 &I_m  &  &  &0&   \\ 
 &  &  &  &   \\ 
 &-\Sigma_{mk}\Sigma_{kk} ^{-1} &  &  &I_k &  
\end{bmatrix}=\begin{bmatrix}
 &\Sigma_{mm}-\Sigma_{mk}\Sigma_{kk}^{-1}\Sigma_{km}  &  &  &0&   \\ 
 &  &  &  &   \\ 
 &0&  &  &\Sigma_{kk} &  
\end{bmatrix}
$$

* เห็นได้ว่า W และ $X_k$ เป็นอิสระต่อกัน (independent) ดังนั้น **การแจกแจงแบบมีเงื่อนไข  $W|X_k$  จะเหมือนกับการแจกแจงของ W   ** ซึ่งเป็นการแจกแจงปกติ (normal distribution) ด้วยค่าคาดหมาย (mean) เท่ากับ $\mu _m-\Sigma_{mk}\Sigma_{kk}^{-1}\mu_k$ และสามารถสรุปได้ว่า

$$
W|X_k\sim N\left (\mu_m-\Sigma_{mk}\Sigma_{kk}^{-1}\mu_k,\Sigma_{mm}-\Sigma_{mk}\Sigma_{kk}^{-1}\Sigma_{km}                     \right )
$$

* ในขณะเดียวกัน เนื่องจากสิ่งที่เราสนใจคือการแจกแจงแบบมีเงื่อนไขเมื่อทราบ $X_k$ ดังนั้น การบวกเพิ่มด้วยฟังก์ชันของ $X_k$ ข้าไปย่อม**มีผลต่อเช่นเดียวกับการบวกเพิ่มด้วยค่าคงที่ซึ่งมีผลต่อค่าคาดหมาย แต่ไม่มีผลต่อค่าความแปรปรวน**

* เราสามารถคำนวณหา $X_m$ ได้จาก $W+\Sigma_{mk}\Sigma_{kk}^{-1}X_k=X_m$ ดังนั้น การแจกแจงแบบมีเงื่อนไข (conditional distribution) ของ $X_m$ เมื่อทราบ $X_k$ เท่ากับ

$$
X_m|X_k\sim N\left (\mu_m-\Sigma_{mk}\Sigma_{kk}^{-1}\mu_k+\Sigma_{mk}\Sigma_{kk}^{-1}X_k,\Sigma_{mm}-\Sigma_{mk}\Sigma_{kk}^{-1}\Sigma_{km}\right)
$$

### <span style=color:purple>ตัวอย่าง: การคำนวณหาค่าคาดหมายแบบมีเงื่อนไข (conditional mean)</span>

#### <span style=color:purple>Example</span>

เช่นเดียวกับตัวอย่างที่แล้ว สมมุติให้ตัวแปรสุ่ม $X_1$ และ $X_2$ มีการแจกแจงปกติ(normally distributed) ด้วยค่าคาดหมาย $(\mu_1,\mu_2)$ และเมทริกซ์ความแปรปรวนร่วม (variance-covariance matrix)

$$
\Sigma=\begin{bmatrix}
 &\sigma^2_1  &  &  &\sigma_{12}&   \\ 
 &  &  &  &   \\ 
 &\sigma_{12}  &  &  &\sigma^2_2&   
\end{bmatrix}
$$

* ใช้สมการด้านบนดังกล่าว เพื่อคำนวณหาค่าคาดหมายแบบมีเงื่อนไข (conditional mean) ของ $X_1$ เมื่อทราบ $X_2=x_2$ ได้เป็น

$$
\begin{align}
E\left [X_1|X_2=x_2\right]&=\mu_1+\Sigma_{12}\Sigma_{22}^{-1}\left ( X_2-\mu_2 \right )
\\
&=\mu_1+\left (\rho\sigma_1\sigma_2\right)\left (\sigma^2_2\right)^{-1}\left (x_2-\mu_2\right)
\\
&=\mu_1+\rho\frac{\sigma_1}{\sigma_2}\left (x_2-\mu_2\right )
\end{align}
$$

* ฟังก์ชันเชิงเส้น (linear function) ของ $x_2$ สังเกตว่าเราไม่ได้กำหนดหรือบังคับให้ค่าคาดหมายแบบมีเงื่อนไข (conditional mean) นี้มีความสัมพันธ์เชิงเส้นแต่อย่างใด แต่คุณสมบัติของการแจกแจงปกตินอกจากนี้ยังสามารถเขียนในรูปของตัวแปรสุ่มได้เป็น

$$
E\left [X_1|X_2\right]=\mu_1+\rho\frac{\sigma_1}{\sigma_2}\left (X_2-\mu_2\right )
$$

### <span style=color:purple>Normal distribution PDF with different standard deviations</span>

```python
## The location (loc) keyword specifies the mean and the scale (scale) keyword specifies the standard deviation

fig, ax = plt.subplots()
x = np.linspace(-10,10,100)
stdvs = [1.0, 2.0, 3.0, 4.0]

for s in stdvs:
    ax.plot(x, norm.pdf(x,scale=s), label='stdv=%.1f' % s)
    
ax.set_xlabel('x')
ax.set_ylabel('pdf(x)')
ax.set_title('Normal Distribution')
ax.legend(loc='best', frameon=True)
ax.set_ylim(0,0.45)
ax.grid(True)
```

### <span style=color:green> Normal distribution PDF with different means</span>

```python
## plot probability distribution functions of normal distribution where the standard deviation is 0 and different means

fig, ax = plt.subplots()
x = np.linspace(-10,10,100)
means = [0.0, 1.0, 2.0, 5.0]

for mean in means:
    ax.plot(x, norm.pdf(x,loc=mean), label='mean=%.1f' % mean)
    
ax.set_xlabel('x')
ax.set_ylabel('pdf(x)')
ax.set_title('Normal Distribution')
ax.legend(loc='best', frameon=True)
ax.set_ylim(0,0.45)
ax.grid(True)
```

### <span style=color:blue>Using random normal distribution values</span>

```python
## create a histogram for the generated numbers

fig, ax = plt.subplots()
xs = norm.rvs(scale=2,size=1000)
x = np.linspace(-10,10,100)
p = norm.pdf(x,scale=2)
v = np.var(xs)
m = np.mean(xs)

ax = fig.add_subplot(111)
ax.hist(xs, bins=10, alpha=0.5, density=True)
ax.plot(x,p, 'r-', lw=2)
ax.set_xlabel('x')
ax.set_ylabel('pdf(x)')
ax.set_title(f'mean={m:.2f}, var={v:.2f}')
ax.grid(True)
```

### <span style=color:orange> A cumulative normal distribution function</span>

```python
## Using fill_between(x, y1, y2=0), it will fill up the area between two curves y1 and y2 which has the default value of 0

fig, ax = plt.subplots()

# for distribution curve
x= np.arange(-4,4,0.001)
ax.plot(x, norm.pdf(x))
ax.set_title("Cumulative normal distribution")
ax.set_xlabel('x')
ax.set_ylabel('pdf(x)')
ax.grid(True)

# for fill_between
px=np.arange(-4,1,0.01)
ax.set_ylim(0,0.5)
ax.fill_between(px,norm.pdf(px),alpha=0.5, color='g')

# for text
ax.text(-1,0.1,"cdf(x)", fontsize=20)
plt.show()
```

### <span style=color:magenta> Calculating the probability of normal distribution </span>

```python
## In norm.cdf, the location (loc) keyword specifies the mean and the scale (scale) keyword specifies the standard deviation

from scipy.stats import norm
lessthan2=norm.cdf(x=2, loc=3, scale=2)
print(lessthan2)
```

```python
## Let’s plot a graph.

fig, ax = plt.subplots()
# for distribution curve
x= np.arange(-4,10,0.001)
ax.plot(x, norm.pdf(x,loc=3,scale=2))
ax.set_title("N(3,$2^2$)")
ax.set_xlabel('x')
ax.set_ylabel('pdf(x)')
ax.grid(True)

# for fill_between
px=np.arange(-4,2,0.01)
ax.set_ylim(0,0.25)
ax.fill_between(px,norm.pdf(px,loc=3,scale=2),alpha=0.5, color='g')

# for text
ax.text(-0.5,0.02,round(lessthan2,2), fontsize=20)
plt.show()
```

```python
## To find the probability of 𝑃(𝑋>4), we can use sf which is called the survival function and it returns 1-cdf. 
## For example, norm.sf(x=4, loc=3, scale=2 returns the probability which is greater than 𝑥=4,𝑃(𝑋>4) when 𝜇=4,𝜎=2.

gr4sf=norm.sf(x=4, loc=3, scale=2)
gr4sf
```

```python
## plot a graph.

fig, ax = plt.subplots()
x= np.arange(-4,10,0.001)
ax.plot(x, norm.pdf(x,loc=3,scale=2))
ax.set_title("N(3,$2^2$)")
ax.set_xlabel('x')
ax.set_ylabel('pdf(x)')
ax.grid(True)
px=np.arange(4,10,0.01)
ax.set_ylim(0,0.25)
ax.fill_between(px,norm.pdf(px,loc=3,scale=2),alpha=0.5, color='g')
ax.text(4.5,0.02,"sf(x) %.2f" %(gr4sf), fontsize=20)

plt.show()
```

### <span style=color:green> Finding quantiles

#### <span style=color:green> Percent point function

```python
### ppf is the inverse of cdf and it is called the Percent point function. 
### Given the mean of 1 and the standard deviation of 3, we can find the quantile a in 𝑃(𝑋<𝑎)=0.506 by using ppf.


norm.ppf(q=0.506, loc=1, scale=3)
```

```python
### plot a graph

fig, ax = plt.subplots()
x= np.arange(-10,10,0.001)
ax.plot(x, norm.pdf(x,loc=1,scale=3))
ax.set_title("N(1,$3^2$)")
ax.set_xlabel('x')
ax.set_ylabel('pdf(x)')
ax.grid(True)
xpoint=norm.ppf(q=0.506, loc=1, scale=3)
px=np.arange(-10,xpoint,0.01)
ax.set_ylim(0,0.15)
ax.fill_between(px,norm.pdf(px,loc=1,scale=3),alpha=0.5, color='g')
ax.text(.8,0.02,"x= %.2f" %xpoint, fontsize=20)
ax.text(-5,0.05,"P(X)=0.506", fontsize=20)

plt.show()
```

#### <span style=color:green>Interval around the mean

### <span style=color:green>Multivariate Normal Distribution

```python
## The multivariate normal distribution is often used to describe any set of correlated real-valued random variables.
## We use multivariate_normal which requires the array of mean and covariance matrix. 
## To make it simple, we use a diagonal matrix in which all off-diagonal elements are zero

## plot a graph

from scipy.stats import multivariate_normal
x,y = np.meshgrid(np.linspace(-10,10,100),np.linspace(-10,10,100))
pos = np.dstack((x,y))
mean = np.array([1, 2])
cov  = np.array([[3,0],[0,15]])
rv = multivariate_normal(mean,cov)
z = rv.pdf(pos)
fig = plt.figure()
ax = fig.add_subplot(111,aspect='equal')
ax.contourf(x,y,z)
ax.set_xlim(-10,10)
ax.set_ylim(-10,10)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('pdf')
```

```python
## We can create a 3D graph using matplotlib’s mpl_toolkits.mplot3d.Axes3D. We also use Scipy frozen RV object.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
from mpl_toolkits.mplot3d import Axes3D

# Create grid and multivariate normal
x = np.linspace(-10,10,500)
y = np.linspace(-10,10,500)
X, Y = np.meshgrid(x,y)
pos = np.empty(X.shape + (2,))
pos[:, :, 0] = X 
pos[:, :, 1] = Y

# Create a frozen RV object
mean = np.array([1, 2])
cov  = np.array([[3,0],[0,15]])
rv = multivariate_normal(mean,cov)

# Make a 3D plot
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot_surface(X, Y, rv.pdf(pos),cmap='viridis',linewidth=0)
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
plt.show()
```
