# <center>การทดสอบสมมติฐาน (Hypothesis Testing)

## บทนำ

หัวข้อนี้นำเสนอหลักการและวิธีการที่เกี่ยวข้องกับการทดสอบสมมุติฐาน โดยให้ความสำคัญกับหลักการต่อไปนี้

* ความสัมพันธ์ระหว่าง Type I and Type II errors และฟังก์ชันอำนาจทางสถิติ (power function)


* บทบาทของ ฟังก์ชันอำนาจทางสถิติ (power function) ในการกำหนดระดับนัยสำคัญทางสถิติ (significance level) และการคำนวณหาค่า p-value


* การทดสอบสมมุติฐานโดยใช้สัดส่วนความเป็นไปได้ (Likelihood Ratio Tests)

## นิยาม: สมมุติฐาน (hypothesis)

**Definition**
สมมุติฐาน (hypothesis) คือข้อความที่เขียนในรูปของค่าพารามิเตอร์ของประชากรหรือค่าพารามิเตอร์ที่แท้จริง (true parameters) ซึ่งแทนด้วย $\theta \ \epsilon \ \Omega  โดยที่\ \Omega$ คือปริภูมิของพารามิเตอร์ (parameter space) หรือเซ็ตของค่าพารามิเตอร์ที่เป็นไปได้ทั้งหมด


* ตัวอย่างเช่น ภายใต้สมมุติฐานว่าง (null hypothesis) ที่ว่า $H_{0} : \lambda  = \lambda _{0} โดยที่ \lambda$  คือพารามิเตอร์ที่แท้จริงของการแจกแจงแบบ Poisson เราจะทราบทันที่ว่ากลุ่มตัวอย่างถูกสุ่มมาจากการแจกแจงแบบ Poisson ที่มีค่าพารามิเตอร์เท่ากับ $\lambda_0$


## นิยาม: สมมุติฐานอย่างง่าย (simple hypothesis) และ สมมุติฐานประกอบ (composite hypothesis)

**Definition**

สมมุติฐานถูกแบ่งออกเป็น 2 สมมุติฐาน ด้วยการแบ่งส่วน (partition) $\Omega$ ออกได้เป็น 2 เซ็ตที่ไม่มีส่วนร่วมต่อกัน (disjoint) $\Omega_0 และ \Omega_1$ ดังต่อไปนี้

1. สมมุติฐานว่าง (null hypothesis) คือสมมุติฐานที่เป็นจริงหากค่าพารามิเตอร์$\theta \ \epsilon \ \Omega_0$ ซึ่งแทนด้วย $H_0 : \theta \ \epsilon \ \Omega_0$

2. สมมุติฐานทางเลือก (alternative hypothesis) คือสมมุติฐานที่เป็นจริงหากค่าพารามิเตอร์ $\theta \ \epsilon \ \Omega_1$ ซึ่งแทนด้วย $H_1 : \theta \ \epsilon \ \Omega_1$

**Definition**

ถ้า $\Omega_i$ มีสมาชิกเพียงตัวเดียว (singleton) เราจะเรียกสมมุติฐาน $H_i$ ว่าเป็นสมมุติฐานอย่างง่าย(simple hypothesis) แต่ถ้า $\Omega_i$ มีสมาชิกมากกว่าหนึ่งตัว 

เราจะเรียกสมมุติฐาน  $H_i$ ว่าเป็นสมมุติฐานประกอบ (composite hypothesis)

### ตัวอย่างสมมุติฐานประกอบ (composite hypothesis)

กองทุน K-USXNDQ มีอัตราผลตอบแทนรายเดือนเท่ากับ $R_A$ ซึ่งมีค่าคาดหมาย (mean) เท่ากับ $µ_A$ ดังนั้น ปริภูมิพารามิเตอร์ (parameter space) 

ในกรณีนี้เท่ากับ $\Omega$ = {$\mu_A \epsilon R$} สมมุติว่าเราต้องการตัดสินใจว่าจะลงทุนในกองทุนนี้หรือไม่

โดยมีเงื่อนไขว่า จะลงทุนในกองทุนนี้ หากค่าคาดหมายของอัตราผลตอบแทนมีค่ามากกว่าหรือเท่ากับร้อยละ 1 ต่อเดือน หรือ $\mu_A \geq  0.01$ ดังนั้น สมมุติฐานที่เหมาะสมในกรณีคือ

<div align="right">(1)</div>
$$H_0 : \mu_A \geq  0.01 \ และ\  H_1 : \mu_A <  0.01$$

ซึ่งในกรณีนี้ $\Omega_0$ = {$(\mu_A,\mu_B)\ \epsilon\ \mathbb{R}^2 \ : \ \mu_A \geq \mu_B$}  และ  $\Omega_1$ = {$(\mu_A,\mu_B)\ \epsilon\ \mathbb{R}^2 \ : \ \mu_A < \mu_B$} และจะเห็นได้ว่า สมมุติฐานทั้งสองอันเป็นสมมุติฐานประกอบ (composite hypothesis)

* กรณีนี้เป็นการสมมุติว่า การตัดสินใจลงทุนขึ้นอยู่กับค่าคาดหมายของอัตราผลตอบแทนเพียงอย่างเดียว โดยไม่สนใจระดับความเสี่ยงของกองทุน ซึ่งในชีวิตจริงไม่ควรทำอย่างยิ่ง เพราะความเสี่ยงมีความสำคัญมากและเป็นเหตุผลสำคัญที่ทำให้การวิเคราะห์ทางการเงินมีความยุ่งยากและซับซ้อน

## ส่วนวิกฤติ (critical region): การเชื่อมโยงสมมุติฐานกับข้อมูล

กำหนดให้ตัวอย่างสุ่ม (random sample) คือข้อมูลที่ได้จากการสุ่ม $X = (X_1, . . . , X_n)$ และปริภูมิตัวอย่าง (sample space) S หมายถึงเซ็ตของค่าของตัวอย่างสุ่มที่เป็นไปได้ทั้งหมด:

<div align="right">(2)</div>
$$S ={X = (X_1, . . . , X_n) : x_i เป็นค่าที่เป็นไปได้ของ X_i สำหรับ i = 1, . . . ,n}$$

โดยที่ ค่าที่เป็นไปได้ของตัวแปรสุ่ม Xi คือค่าจำนวนจริงที่อยู่ในส่วนค้ำจุน (support) ของตัวแปรสุ่ม $X_i$ นั่นเอง

**Definition**

ส่วนวิกฤติ (critical region) ของกระบวนการทดสอบ

$S_1$ ={$x\ \epsilon \ S : ผลการทดสอบสมมุติฐานโดยใช้ข้อมูล x ระบุว่าปฏิเสธ H_0$} หมายถึง เซ็ตของค่าของตัวอย่างสุ่มที่เมื่อนำมาทดสอบสมมุติฐานแล้วจะสรุปได้ว่าปฏิเสธ $H_0$

## ส่วนวิกฤติ (critical region) ในรูปของค่าสถิติ (Statistic)
* เพื่อความสะดวกในการวิเคราะห์ เรามักนิยามส่วนวิกฤติ (critical region) โดยใช้ค่าสถิติสำหรับการทดสอบสมมุติฐาน (test statistic) T = r(X)

**Definition**

กำหนดให้X คือตัวอย่างสุ่มที่สุ่มมาจากการแจกแจงที่ขึ้นอยู่กับพารามิเตอร์$\theta $ และสมมุติฐาน

<div align="right">(3)</div>
$$H_{0} : \theta \ \epsilon \ \Omega_{0} และ H_{1} :\theta \ \epsilon \ \Omega_{1},$$

ส่วนวิกฤติ (critical region) $S_{1} = {x\ \epsilon\ S : r(x)\ \epsilon \ R}$ ซึ่งเขียนในรูปของค่าสถิติ T

### ตัวอย่าง: ส่วนวิกฤติ (critical region) ในรูปของค่าสถิติ (Statistic)
* พิจารณาการทดสอบว่าค่าคาดหมาย $\mu มีค่าเท่ากับค่าจำนวนจริง \mu_{0}$ หรือไม่? นั่นคือ $H_{0} : µ = µ_{0}$

* การทดสอบนี้สามารถทำได้โดยใช้ค่าสถิติ $T = |\bar{X}_{n} − µ_{0}| โดยที่  \bar{X}_{n} = \frac{\sum_{i=1}nX_{i}}{n}$ และกำหนดให้ส่วนวิกฤติคือช่วง R = [0, $\infty $)

* ดังนั้น เราจะปฏิเสธ $H_{0} : µ = µ_{0}$ ถ้าค่าสถิติที่คำนวณได้จากข้อมูลอยู่ในช่วง [0,$\infty $)

* สมมุติฐานอันเดียวกันนี้สามารถทดสอบได้โดยใช้ค่าสถิติที่ต่างออกไป เช่น $T = \frac{\bar{X}_{n}}{\mu_{0}}$


คำถามที่สำคัญก็คือ แล้วเราควรจะเลือกใช้กระบวนการทดสอบแบบไหนดี? คำตอบก็คือ ควรจะเลือกวิธีที่มีโอกาสผิดพลาดน้อยที่สุด

## นิยาม:ความผิดพลาดของการทดสอบสมมุติฐาน

**Definition**

การทดสอบสมมุติฐานสามารถเกิดความผิดพลาดได้ 2 รูปแบบดังต่อไปนี้

* ความผิดพลาดประเภทที่ 1 (Type I error) หมายถึงความผิดพลาดที่เกิดขึ้นจากการปฏิเสธ $H_{0} ทั้งที่ในความเป็นจริงสมมุติฐานนั้นเป็นจริงนั่นคือ\  \theta \ \epsilon \ \Omega _{0} $


* ความผิดพลาดประเภทที่ 2 (Type II error) หมายถึงความผิดพลาดที่เกิดขึ้นจากการไม่ปฏิเสธ$H_{0} ทั้งที่ในความเป็นจริงสมมุติฐานนั้นไม่เป็นจริงนั่นคือ\  \theta \ \epsilon \ \Omega _{1} $

## นิยาม: ฟังก์ชันอำนาจทางสถิติ (power function) ของกระบวนการทดสอบ δ

**Definition**
ฟังก์ชันอำนาจทางสถิติ (power function) ของกระบวนการทดสอบ δ ที่กำหนดให้ส่วนวิกฤติคือ $S_{1}$ สามารถเขียนได้เป็น

<div align="right">(4)</div>
$$ \pi \ (\theta | \delta ) = Pr(X \ \epsilon \ S_{1} | \theta ), \forall \theta \ \epsilon \ \Omega$$
    
หรือ 

<div align="right">(5)</div>
$$ \pi \ (\theta | \delta ) = Pr(T(X) \ \epsilon \ R | \theta ), \forall \theta \ \epsilon \ \Omega$$

ฟังก์ชันอำนาจทางสถิติ (power function) เป็นฟังก์ชันของพารามิเตอร์ที่แท้จริงหรือค่าพารามิเตอร์ทางทฤษฎี $\theta$ และขึ้นอยู่กับการแจกแจงของตัวอย่างสุ่มที่ขึ้นอยู่กับค่าพารามิเตอร์ $\theta$ นอกจากนี้ ยังขึ้นอยู่กับกระบวนการทดสอบผ่านช่องทางส่วนวิกฤติ$S_{1}$ หรือ หากพิจารณาในรูปของค่าสถิติ T ก็จะขึ้นอยู่กับส่วนวิกฤติ R

### ตัวอย่าง: ฟังก์ชันอำนาจทางสถิติ (power function) ของกระบวนการทดสอบ δ

* พิจาณาการทดสอบสมมุติฐาน

$$H_{0} : \mu = \mu_{0} \ และ\  H_{1}: \mu \neq \mu_{0} $$

โดยทราบว่าค่าความแปรปรวนเท่ากับ $\sigma^{2}$
* พิจารณากระบวนการทดสอบ δ ที่ทดสอบโดยใช้ค่าสถิต $T = |\bar{X}_{n} − µ_{0}|$ และมีส่วนวิกฤติ R = [0, $\infty $)

* คำถามในที่นี้คือ ฟังก์ชันอำนาจทางสถิติ (power function) ของกระบวนทดสอบนี้เท่ากับเท่าใด?

* เนื่องจากตัวแปรสุ่ม $X_{i}$ มีการแจกแจงแบบปกติที่มีค่าคาดหมายเท่ากับ $\mu$ และค่าความแปรปรวนเท่ากับ $\sigma_{2}$ ทำให้สามารถสรุปได้ว่า ค่าคาดหมายของตัวอย่างสุ่ม $(sample \ mean) \bar{X}_{n}$ มีการแจกแจงแบบปกติที่มีค่าคาดหมายเท่ากับ $\mu$ และค่าความแปรปรวนเท่ากับ $\frac{\sigma^{2}}{n}$

### ตัวอย่าง: ฟังก์ชันอำนาจทางสถิติ (power function) สำหรับการแจกแจงปกติ
* ดังนั้น ฟังก์ชันอำนาจทางสถิติ (power function) สำหรับค่าพารามิเตอร์ $\mu$ ใดๆ เท่ากับ

$ \pi (\mu |\delta ) = Pr(T(X) \ \epsilon R|\mu) = Pr(|\bar{X}_{n} − \mu_{0}|\ \epsilon \ [c, \infty) | \mu)\\ 
= Pr(|\bar{X}_{n} \geq  \mu_{0} + c| \mu ) + Pr(|\bar{X}_{n} \leq   \mu_{0} - c| \mu ) \\
=Pr(\frac{\bar{X}_{n} - \mu}{\frac{\sigma }{\sqrt{n}}} \geq \frac{\mu_{0}+c-\mu}{\frac{\sigma }{\sqrt{n}}}|\mu) +Pr(\frac{\bar{X}_{n} - \mu}{\frac{\sigma }{\sqrt{n}}} \geq \frac{\mu_{0}-c-\mu}{\frac{\sigma }{\sqrt{n}}}|\mu) \\
= 1- Φ \frac{\mu_{0}+c-\mu}{\frac{\sigma }{\sqrt{n}}} + Φ\frac{\mu_{0}-c-\mu}{\frac{\sigma }{\sqrt{n}}} $

* สังเกตว่า ฟังก์ชันอำนาจทางสถิติ (power function) $\pi$ (µ|δ) เป็นฟังก์ชันของ µ และค่าของฟังก์ชันนี้ขึ้นอยู่กับค่า c ซึ่งเป็นค่าที่กำหนดรูปแบบของกระบวนการทดสอบ และค่า $µ_0$ ซึ่งเป็นค่าที่กำหนดรูปแบบของสมมุติฐาน นั่นคือ ค่าของฟังก์ชันอำนาจทางสถิติ(power function) ขึ้นอยู่กับทั้งรูปแบบของสมมุติฐานและรูปแบบของกระบวนการทดสอบ

## ฟังก์ชันอำนาจทางสถิติ (power function) ของกระบวนการทดสอบ δ

รูปด้านล่างแสดงค่าฟังก์ชันอำนาจทางสถิติ (power function) สำหรับการทดสอบสมมุติฐานที่ กำหนดให้$\mu_0$ = 3 ด้วยข้อมูลที่มีทั้งหมด n = 1,000 ตัวอย่าง 

และสมมุติให้ $\sigma^2$ = 10 สำหรับกระบวนการทดสอบที่แตกต่างกันซึ่งกำหนดด้วยค่า c = 0.3, 0.5, 0.7

ฟังก์ชันอำนาจทางสถิติ(power function)ในอุดมคติ

$$ \pi \ (\theta | \delta ) = \;\;1\;\; สำหรับทุกๆค่า \;\;θ ∈ Ω1$$
$$ \pi \ (\theta | \delta ) = \;\;0\;\; สำหรับทุกๆค่า \;\;θ ∈ Ω0$$
    
* ความยุ่งยากในการทดสอบสมมุติฐานเกิดขึ้นจากการที่ระดับความผิดพลาดประเภทที่1 (Type I error) และระดับความผิดพลาดประเภทที่2(Type II error) มักจะมีค่าที่สวนทางกัน 
    
* ทางออกคือการเลือกกระบวนการทดสอบที่มีความสมดุลระหว่างความผิดพลาดทั้งสอง ประเภท

## ระดับนัยสำคัญทางสถิติ(StatisticalSigniﬁcanceLevel) 

วิธีการที่ได้รับความนิยมอันหนึ่งคือการเลือกจำนวนจริง $α0 ∈ (0,1)$ ที่ทำให้ $TypeI error$ สำหรับทุกๆค่า $θ ∈ Ω0$ มีค่าไม่เกิน $α0$ ซึ่งหมายถึงการกำหนดให้ Type I errorมีค่าไม่เกิน $α0$ 

กระบวนการทดสอบ(test)ที่สอดคล้องกับเงื่อนไข 

$$π (θ|δ) ≤ α0, สำหรับทุกค่า θ ∈ Ω0 (6)$$

ถูกเรียกว่าการทดสอบที่ระดับ $α0$ กล่าวคือการทดสอบนี้มีระดับนััยสำคัญ(signiﬁcancelevel) เท่ากับ $α0$ 

ส่วนขนาดการทดสอบ(size of test)ของกระบวนการทดสอบ $δ$ สามารถนิยามได้เป็น

$$α(δ) = sup θ∈Ω0 π (θ|δ) (7)$$

* ถ้าขนาดการทดสอบมีค่าเท่ากับ $α0$ แล้วกระบวนการทดสอบ $δ$ นั้นจะมีระดับนััยสำคัญ มากกว่าหรือเท่ากับ $α0$ เสมอ 

## การออกแบบกระบวนการทดสอบสมมติฐานให้มีระดับนัยสำคัญที่ต้องการ

สมมุติฐานที่ต้องการทดสอบคือ

$$H0 : θ ∈ Ω0$$  
และ  $$H1 : θ ∈ Ω1$$

* กระบวนการทดสอบที่พิจารณาใช้่าสถิติTและส่วนวิกฤติ $R = [c,∞)$ 
* ดังนั้นการออกแบบกระบวนการทดสอบในที่นี้คือการเลือกTและcที่ทำให้

$$ ∑θ ∈Ω0 Pr(T≥c|θ) ≤ α0 \;\;\; (8)$$

## ระดับความน่าจะเป็นที่จะปฏิเสธสมมุติฐานจริง (p-value)

* ค่า p-value คือค่าระดับนัยสำคัญทางสถิติ $\underline{\alpha_0 }$ ที่ต่ำที่สุดที่การทดสอบโดยใช้ข้อมูลที่มีอยู่ (observed data) สามารถปฏิเสธสมมุติฐาน $H_0$ ได้ที่ระดับนัยสำคัญทางสถิตินั้น ซึ่งหมายความว่า ย่อมจะปฏิเสธสมมุติฐาน $H_0$ สำหรับการทดสอบสมมุติฐานด้วยข้อมูลอันเดียวกันที่ระดับนัยสำคัญทางสถิติ $\alpha_0\geq \underline{\alpha_0 }$ ใดๆ
|
### Definition 

ระดับความน่าจะเป็นที่จะปฏิเสธสมมุติฐานจริง (p-value) หมายถึงระดับนัยสำคัญทางสถิติที่ต่ำที่สุด $\underline{\alpha_0 }\equiv \hat{p}$ ที่การทดสอบโดยใช้ข้อมูลที่มีอยู่(observed data) สามารถปฏิเสธสมมุติฐาน $H_0$ ได้ที่ระดับนัยสำคัญทางสถิตินั้น

* นักทดลอง (experimenter) ที่ต้องการการทดสอบที่ระดับ $\alpha_0$ จะปฏิเสธสมมุติฐานว่างก็ต่อเมื่อ $p-value,\hat{p}\left ( X|\delta  \right )$ , มีค่าไม่เกิน $\alpha_0$ บางครั้งจึงเรียก p-value ว่า ระดับนัยสำคัญทางสถิติที่คำนวณได้(observed significance level)

### ตัวอย่าง: ระดับความน่าจะเป็นที่จะปฏิเสธสมมุติฐานจริง (p-value)
* พิจารณากระบวนการทดสอบสมมุติฐาน

$$
H_0:\mu=\mu_0 \,และ\, H_1:\mu\neq \mu_0
$$

ที่ใช้ค่าสถิติ $Z=\frac{\tilde{X}_n-\mu_0}{\frac{\sigma }{\sqrt{n}}} $

* สมมุติว่าค่าสถิติZ จากข้อมูลที่เก็บรวบรวมมาได้มีค่าเท่ากับ $\hat{Z}=-3.14$ ซึ่งทำให้สรุปได้ว่า เราควรจะต้องปฏิเสธ $H_0$ ที่ระดับนัยสำคัญ $\alpha_0$ ที่สอดคล้องกับเงื่อนไข

$$
\left | \hat{Z} \right |=\left | -3.14 \right |\geq \Phi ^{-1}\left (1-\frac{\alpha_0}{2}\right )
$$

ซึ่งคำตอบที่ได้ก็คือ $\underline{\alpha_0 }=0.0017$ ดังนั้น จึงสามารถสรุปได้ว่า ระดับความน่าจะเป็นที่จะปฏิเสธสมมุติฐานจริง (p-value) $\hat{p}=0.0017$

รูปด้านล่างแสดงค่า p-value โดยเริ่มจากการคำนวณค่า $\hat{Z}$ แล้วจึงคำนวณหาค่า p-value จากพื้นที่ใต้กราฟที่แรเงาด้วยสีแดง

## การทดสอบสมมุติฐานโดยใช้สัดส่วนความเป็นไปได้ (Likelihood Ratio Tests)

* ค่าสถิติ(statistic) ที่ได้รับความนิยมอันหนึ่งคือ ค่าสัดส่วนความเป็นไปได้ (likelihood ratio statistic) 

$$
\Lambda \left ( x \right )=\frac{sup_{\theta \varepsilon\Omega_{0}}{\mathscr{L}}\left (\theta |x \right )}{sup_{\theta \varepsilon\Omega}{\mathscr{L}}\left (\theta |x \right )}
$$

* ค่าสัดส่วนความเป็นไปได้(likelihood ratio statistic) สามารถนำมาใช้ทดสอบสมมุติฐานที่อยู่ในรูปแบบ

$$
H_{0}:\theta \, \epsilon \, \Omega_{0} และ H_{1}:\theta \, \epsilon \, \Omega_{1}
$$

โดยที่จะปฏิเสธสมมุติฐาน $H_0$ ถ้า $\Lambda(x)\leq \lambda$ สำหรับค่าคงที่ $\lambda$ ค่าใดค่าหนึ่ง

* การทดสอบสมมุติฐานด้วยวิธีนี้จะปฏิเสธ $H_0$ ก็ต่อเมื่อค่าสถิติ $\Lambda(x)$ มีค่าต่ำ ซึ่งหมายความว่าข้อจำกัดที่มาจาก $H_0$ มีผลทำให้ค่าสูงสุดของฟังก์ชันความเป็นไปได้มีค่าลดลงอย่างมากเมื่อเทียบกับกรณีที่ไม่มีข้อจำกัด ซึ่งมีนัยว่า ค่าพารามิเตอร์ที่แท้จริงน่าจะไม่อยู่ใน $\Omega_{0}$ จึงควรปฏิเสธสมมุติฐาน

## ตัวอย่าง: การทดสอบสมมุติฐานโดยใช้สัดส่วนความเป็นไปได้(Likelihood Ratio Tests)

* กำหนดให้ $X_i$ คือตัวแปรสุ่มที่แทนผลของการทดลองยาชนิดใหม่ซึ่งมีพารามิเตอร์ $\theta$ แทนความน่าจะเป็นของความสำเร็จในแต่ละครั้ง

* สมมุติว่า Y คือตัวแปรสุ่มที่แทนจำนวนครั้งที่จะสำเร็จ

* ฟังก์ชันความเป็นไปได้สำหรับการทดลองทั้งหมด n ครั้งในรูปของจำนวนครั้งที่สำเร็จคือ

$$
\mathscr{L}\left ( \theta |y \right ) = \frac{n!}{y!(n-y)!}\theta ^{y}({1-\theta})^{n-y}
$$

* ตัวอย่างนี้ต้องการทดสอบ

$$
H_0 : \theta = \theta_0 และ H_1 : \theta = \theta_0
$$

* ในกรณีนี้ $\Omega_0 = \left \{ \theta _0 \right \}$ และ $\Omega=[0,1]$ ดังนั้น ค่าสัดส่วนความเป็นไปได้(likelihood ratio statistic) เท่ากับ

$$
\Lambda_y = \frac{\theta_0^y(1-\theta_0)^{n-y}}{sup_{\theta \epsilon[0,1]}\theta^y(1-\theta)^{n-y}}
$$

* คำตอบของปัญหาการหาค่าความเป็นไปได้สูงสุดในส่วนของตัวหารคือ ตัวประมาณค่า MLE

$$
\hat{\theta}_{MLE}(y) = \frac{y}{n}
$$

และค่าฟังก์ชันความเป็นไปได้สูงสุดเท่ากับ

$$
\left(\frac{y}{n}\right )^{y}\left (\frac{n-y}{n}\right)^{n-y}
$$

* ดังนั้น ค่าสถิติ $\Lambda (y)$ มีค่าเท่ากับ 

$$
\begin{align}
\Lambda (y) &= \frac{\theta_0^y(1-\theta_0)^{n-y}}{\left (\frac{y}{n}  \right )^y\left ( \frac{n-y}{n} \right )^{n-y}}
\\
&=\left (\frac{n\theta_{0} }{y}\right )^{y}\left (\frac{n(1-\theta_0 )}{n-y}\right )^{n-y}
\end{align}
$$

* สมมุติเพิ่มเติมอีกว่า มีการทดลองทั้งหมด n = 10 ครั้ง และค่า $\theta_0=0.6$

* คำถามแรกที่ต้องการตอบคือ ถ้าระดับนัยสำคัญทางสถิติ $\alpha _0=0.05$ เราควรจะต้องเลือกค่าคงที่ $\lambda$ เท่ากับเท่าใด?

* การหาคำตอบในส่วนนี้เริ่มจากนิยามของระดับนัยสำคัญทางสถิติ

$$
\begin{align}
\alpha _0\geq \sup_{\theta \epsilon \Omega _0}^{}\pi \left ( \theta |\delta  \right )&=\pi \left ( \theta _0| \delta\right )
\\
&=Pr\left ( \Lambda (y) \leq \lambda |\theta _0\right )
\\
&=Pr\left ( Y|\theta_0\right )
\end{align}
$$

โดยที่เซ็ต $Y = \left \{ y\, \epsilon\, \left \{ 0,1,...,10 \right \}:\Lambda (y) \leq \lambda  \right \}$ และ

* เนื่องจากการทดลองแต่ละครั้งเป็นอิสระต่อกัน ดังนั้น $Pr(Y|\theta_0)$ จะมีค่าเท่ากับผลรวมของความน่าจะเป็นของแต่ละค่า $y\, \epsilon \, Y$

* ดังนั้น วิธีการหาค่า $\lambda$ สามารถทำได้ด้วยการคำนวณหาค่า $\Lambda (y)$ และ $Pr\left ( y|\theta_0\right )$ สำหรับแต่ละค่าของ y ที่เป็นไปได้และนำมาเรียงลำดับจากค่า y ที่ให้ค่า $\Lambda (y)$ จากต่ำสุดไปถึงค่าสูงสุด ดังแสดงในตาราง

* หลังจากนั้นจึงหาค่าผลรวมของ $Pr\left ( y|\theta_0\right )$ จากตำแหน่งต่ำสุดไปเรื่อยๆ จนได้ค่าที่ใกล้เคียงแต่ไม่เกิน $\alpha_0=0.5$ ซึ่งในที่นี้มีค่าเท่ากับ 0:018 และ ผลบวกนี้จบลงที่ตำแหน่งที่ค่า $\Lambda (y)=0.035$

* ดังนั้น ค่าคงที่ $\lambda$ ที่ต้องการคือ ค่าจำนวนจริงใดๆ ที่อยู่ระหว่าง [ 0.035 , 0.104 )

<img src="images/14.5.jpg" />

* ขนาดของการทดสอบเท่ากับ $\alpha\left ( \delta  \right )=\sup_{\theta \epsilon \Omega _0}^{}\pi \left ( \theta |\delta  \right )=Pr\left ( Y|\theta _0 \right )=0.018$

* สังเกตว่า เราจะปฏิเสธ $H_0$ สำหรับค่า y = 0; 1; 2 หรือ 10

##  ทฤษฎี: การทดสอบสมมุติฐานโดยใช้สัดส่วนความเป็นไปได้(Likelihood Ratio Tests) โดยใช้การแจกแจงที่ลิมิต (Asymptotic Distribution)

### Theorem

กำหนดให้ $\Omega$ เป็นปริภูมิของพารามิเตอร์ $\theta$ ที่ประกอบไปด้วย p พารามิเตอร์และสมมุติให้เงื่อนไขที่จำเป็นสำหรับการพิสูจน์ว่าการประมาณค่าด้วยความเป็นไปได้สูงสุด (MLE) มีการแจกแจงที่ลิมิตเป็นการแจกแจงปกติพิจารณาสมมุติฐาน $H_0$ เป็นสมมุติฐานที่กำหนดให้ $h_i(\theta )=\bar{h_i}$ สำหรับ i = 1,...,k โดยที่ $\bar{h_i}$ คือค่าคงที่ดังนั้น เราสามารถพิสูจน์ได้ว่า ภายใต้ข้อกำหนดของ $H_0$

$$
-2log\Lambda \left ( X \right )\overset{d}{\rightarrow}X^2_k,
$$

* ทฤษฎีบทด้านบนช่วยให้เราสามารถทราบการแจกแจงของ $-2log\Lambda \left ( X \right )$ ได้โดยที่ไม่ต้องคำนวณอะไรที่ยุ่งยาก

* ดังนั้น เราจึงนิยมทดสอบสมมุติฐานโดยใช้ค่าสถิติ $-2log\Lambda \left ( X \right )$ มากกว่า $\Lambda \left ( X \right )$

เราจะปฏิเสธสมมุติฐาน $H_0$ ถ้า $-2log\Lambda \left ( X \right ) \geq  \bar{\lambda }$

##  รูปภาพ: การทดสอบสมมุติฐานโดยใช้สัดส่วนความเป็นไปได้(Likelihood Ratio Tests) โดยใช้การแจกแจงที่ลิมิต (Asymptotic Distribution)

* เราสามารถกำหนดค่าวิกฤติ $\bar\lambda$ สำหรับการทดสอบที่มีระดับนัยสำคัญ $\alpha_0$ ใดๆ ได้โดยใช้ $-2log\Lambda\left(X\right)$ $\geq$ $\left(X^2\right)^{-1}\left(1-\alpha_0\right)$

* สังเกตได้ว่าการทดสอบโดยใช้การแจกแจงแบบไคว์กำลังสอง (Chi square) จะใช้พื้นที่ส่วนท้ายของการแจกแจง (tail area) เพียงด้านเดียว เนื่องจากการแจกแจงแบบไคว์กำลังสองมีส่วนค้ำจุน (support) ที่เป็นบวกเท่านั้น
